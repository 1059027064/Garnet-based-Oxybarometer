```python
# Relative molecular masses of oxides (unit: g/mol) 
molar_masses = {
    'SiO2': 60.08,   # SiO2
    'TiO2': 79.87,   # TiO2
    'Al2O3': 101.96, # Al2O3
    'FeO': 71.84,    # FeO
    'MgO': 40.30,    # MgO
    'CaO': 56.08,    # CaO
    'Na2O': 61.98,   # Na2O
    'Cr2O3': 151.99, # Cr2O3
    'MnO': 70.94     # MnO
}

# Molar masses of elements (unit: g/mol)
element_molar_masses = {
    'Si': 28.0855,  # Si
    'Ti': 47.867,   # Ti
    'Al': 26.9815,  # Al
    'Fe': 55.845,   # Fe
    'Mg': 24.305,   # Mg
    'Ca': 40.078,   # Ca
    'Na': 22.9898,  # Na
    'Cr': 51.996,   # Cr
    'Mn': 54.938    # Mn
}

# Number of metal cations (number of cations in each oxide) 
cation_numbers = {
    'Si': 1, 'Ti': 1, 'Al': 2, 
    'Fe': 1, 'Mg': 1, 'Ca': 1, 
    'Na': 2, 'Cr': 2, 'Mn': 1  
}

# Function to calculate correction factor (only supports garnet, Grt)
def calculate_chemical_formula(df):
    oxygen_atoms = 12  # Fixed number of oxygen atoms in garnet (Grt)
    
    # Define element columns for garnet 
    Elements = ['Si_Grt', 'Ti_Grt', 'Al_Grt', 'Fe_Grt', 
                'Mg_Grt', 'Ca_Grt', 'Na_Grt', 'Cr_Grt', 'Mn_Grt']
    
    results = []

    # Process each row of data
    for idx, row in df.iterrows():
        # Extract mass percentages of each oxide 
        SiO2_mass = row['Si_Grt']
        TiO2_mass = row['Ti_Grt']
        Al2O3_mass = row['Al_Grt']
        FeO_mass = row['Fe_Grt']
        MgO_mass = row['Mg_Grt']
        CaO_mass = row['Ca_Grt']
        Na2O_mass = row['Na_Grt']
        Cr2O3_mass = row['Cr_Grt'] 
        MnO_mass = row['Mn_Grt']  

        # Calculate moles of oxides = mass percentage / molar mass
        n_Si = SiO2_mass / molar_masses['SiO2']
        n_Ti = TiO2_mass / molar_masses['TiO2']
        n_Al = Al2O3_mass / molar_masses['Al2O3']
        n_Fe = FeO_mass / molar_masses['FeO']
        n_Mg = MgO_mass / molar_masses['MgO']
        n_Ca = CaO_mass / molar_masses['CaO']
        n_Na = Na2O_mass / molar_masses['Na2O']
        n_Cr = Cr2O3_mass / molar_masses['Cr2O3']  
        n_Mn = MnO_mass / molar_masses['MnO']      

        # Calculate correction factor z (based on total oxygen atoms)
        total_oxygen = (
            n_Si * 2 +    # SiO2 provides 2 oxygen atoms
            n_Ti * 2 +    # TiO2 provides 2 oxygen atoms
            n_Al * 3 +    # Al2O3 provides 3 oxygen atoms
            n_Fe * 1 +    # FeO provides 1 oxygen atom
            n_Mg * 1 +    # MgO provides 1 oxygen atom
            n_Ca * 1 +    # CaO provides 1 oxygen atom
            n_Na * 1 +    # Na2O provides 1 oxygen atom 
            n_Cr * 3 +    # Cr2O3 provides 3 oxygen atoms 
            n_Mn * 1      # MnO provides 1 oxygen atom 
        )
        z = oxygen_atoms / total_oxygen
        
        # Calculate cation numbers for each element (added Cr and Mn)
        ion_numbers = {
            'Si': n_Si * cation_numbers['Si'] * z,
            'Ti': n_Ti * cation_numbers['Ti'] * z,
            'Al': n_Al * cation_numbers['Al'] * z,
            'Fe': n_Fe * cation_numbers['Fe'] * z,
            'Mg': n_Mg * cation_numbers['Mg'] * z,
            'Ca': n_Ca * cation_numbers['Ca'] * z,
            'Na': n_Na * cation_numbers['Na'] * z,
            'Cr': n_Cr * cation_numbers['Cr'] * z, 
            'Mn': n_Mn * cation_numbers['Mn'] * z  
        }

        # Store results in a dictionary
        result = {
            'Si': ion_numbers['Si'],
            'Ti': ion_numbers['Ti'],
            'Al': ion_numbers['Al'],
            'Cr': ion_numbers['Cr'], 
            'Fe': ion_numbers['Fe'],
            'Mn': ion_numbers['Mn'],
            'Mg': ion_numbers['Mg'],
            'Ca': ion_numbers['Ca'],
            'Na': ion_numbers['Na'],
            'Correction_Factor_z': z
        }
        
        results.append(result)

    # Convert results to DataFrame
    formula_df = pd.DataFrame(results)
    return formula_df
```

```python
def garnet_cation_filter(df_input):

    import pandas as pd

    oxide_cols = [
        'Si_Grt','Ti_Grt','Al_Grt','Cr_Grt',
        'Fe_Grt','Mn_Grt','Mg_Grt','Ca_Grt','Na_Grt'
    ]

    # Automatically identify other (non-oxide) columns
    other_cols = [c for c in df_input.columns if c not in oxide_cols]

    # ===============================
    # 1. Calculate structural formula cations
    # ===============================

    formula_df = calculate_chemical_formula(df_input[oxide_cols])

    # Remove correction factor column if present
    if 'Correction_Factor_z' in formula_df.columns:
        formula_df = formula_df.drop(columns=['Correction_Factor_z'])

    # Ensure all column names end with '_Grt'
    formula_df.columns = [
        c if c.endswith('_Grt') else f"{c}_Grt"
        for c in formula_df.columns
    ]

    # ===============================
    # 2. Preserve original oxide data
    # ===============================

    origin_df = df_input[oxide_cols].add_suffix('_origin')

    # ===============================
    # 3. Keep other columns
    # ===============================

    other_df = df_input[other_cols]

    # ===============================
    # 4. Merge all DataFrames
    # ===============================

    df = pd.concat([
        formula_df,
        origin_df,
        other_df
    ], axis=1)

    # ===============================
    # 5. Rename variables
    # ===============================

    df = df.rename(columns={
        'Y value_P': 'P',
        'Y value_T': 'T',
        'Y value_fO2': 'fO2'
    })

    # ===============================
    # 6. Calculate total cations
    # ===============================

    cation_cols = [
        'Si_Grt','Ti_Grt','Al_Grt','Cr_Grt',
        'Fe_Grt','Mn_Grt','Mg_Grt','Ca_Grt','Na_Grt'
    ]

    df['cation_sum'] = df[cation_cols].sum(axis=1)

    # ===============================
    # 7. Filter by cation sum (7.9–8.1)
    # ===============================

    original_count = len(df)

    df = df[
        (df['cation_sum'] >= 7.9) &
        (df['cation_sum'] <= 8.1)
    ]

    removed_count = original_count - len(df)

    print(f"Removed samples (cation filter): {removed_count}")

    return df
```

```python

# Source code reference: :contentReference[oaicite:0]{index=0}
def run_full_oxybarometer_pipeline(
        df_predict,
        prediction_name,
        df_P_T_train,
        df_fO2_train,
        BASE_FEATURES,
        P_model,
        T_model,
        fO2_model,
        calculate_iw,
        RESULT_DIR,
        RANDOM_STATE=42):

    import numpy as np
    import pandas as pd
    import matplotlib.pyplot as plt
    import os
    from sklearn.preprocessing import StandardScaler
    from sklearn.decomposition import PCA
    from sklearn.neighbors import NearestNeighbors

    print("====================================")
    print("Running full oxybarometer pipeline...")
    print("====================================\n")

    predict_df = df_predict.copy()

    # ==========================================================
    # 1. OOD (Out-of-Distribution) Detection
    # ==========================================================

    # Combine training datasets
    train_full_df = pd.concat(
        [df_P_T_train, df_fO2_train],
        ignore_index=True
    )

    # Extract features
    X_train = train_full_df[BASE_FEATURES]
    X_predict = predict_df[BASE_FEATURES]

    # Standardization
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_predict_scaled = scaler.transform(X_predict)

    # PCA transform for visualization
    pca = PCA(n_components=2, random_state=RANDOM_STATE)
    X_train_pca = pca.fit_transform(X_train_scaled)
    X_predict_pca = pca.transform(X_predict_scaled)

    # Nearest-neighbour distance for OOD detection
    nn = NearestNeighbors(n_neighbors=1)
    nn.fit(X_train_scaled)

    distances, _ = nn.kneighbors(X_predict_scaled)
    distances = distances.flatten()

    # Compute internal training nearest-neighbour distances
    train_nn = NearestNeighbors(n_neighbors=2)
    train_nn.fit(X_train_scaled)

    train_distances, _ = train_nn.kneighbors(X_train_scaled)
    train_nn_dist = train_distances[:, 1]

    # Define OOD threshold as the 99th percentile of training distances
    threshold_ood = np.percentile(train_nn_dist, 99)

    # Classify OOD samples
    ood_mask = distances > threshold_ood

    predict_df["OOD_distance"] = distances
    predict_df["OOD_level"] = "In-Distribution"
    predict_df.loc[ood_mask, "OOD_level"] = "OOD"
    predict_df["is_OOD"] = ood_mask

    print("OOD ratio:",
          round(np.mean(ood_mask) * 100, 2), "%\n")

    # PCA visualization with OOD highlighting
    plt.figure(figsize=(8, 6))

    # Training data
    plt.scatter(
        X_train_pca[:, 0],
        X_train_pca[:, 1],
        alpha=0.3,
        color="gray",
        label="Train"
    )

    # In-Distribution prediction samples
    plt.scatter(
        X_predict_pca[~ood_mask, 0],
        X_predict_pca[~ood_mask, 1],
        alpha=0.7,
        color="blue",
        label="In-Distribution"
    )

    # OOD prediction samples
    plt.scatter(
        X_predict_pca[ood_mask, 0],
        X_predict_pca[ood_mask, 1],
        alpha=0.9,
        color="red",
        edgecolors="black",
        linewidths=0.5,
        label="OOD"
    )

    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.title("Feature Space PCA Projection")
    plt.legend(frameon=False)
    plt.tight_layout()
    plt.show()

    # ==========================================================
    # 2. P–T Prediction
    # ==========================================================

    print("Predicting P and T...")

    X_pred = predict_df[BASE_FEATURES]

    predict_df["P_pred"] = P_model.predict(X_pred)
    predict_df["T_pred"] = T_model.predict(X_pred)

    print("Mean P:", round(predict_df["P_pred"].mean(), 3))
    print("Mean T:", round(predict_df["T_pred"].mean(), 3), "\n")

    # ==========================================================
    # 3. IW Calculation
    # ==========================================================

    use_iw_feature = False
    try:
        n_expected_with_iw = len(BASE_FEATURES) + 1
        if hasattr(fO2_model, "n_features_in_"):
            use_iw_feature = (fO2_model.n_features_in_ == n_expected_with_iw)
    except Exception:
        use_iw_feature = False

    # Calculate IW buffer
    predict_df["IW_pred"] = calculate_iw(
        predict_df["P_pred"],
        predict_df["T_pred"]
    )

    print("Mean IW:",
          round(predict_df["IW_pred"].mean(), 3), "\n")

    # ==========================================================
    # 4. fO2 Prediction
    # ==========================================================

    # Select features depending on model requirement
    if use_iw_feature:
        fO2_features = BASE_FEATURES + ["IW_pred"]
    else:
        fO2_features = BASE_FEATURES

    predict_df["logfO2_pred"] = fO2_model.predict(
        predict_df[fO2_features]
    )

    # Calculate ΔIW
    predict_df["dIW_pred"] = (
        predict_df["logfO2_pred"]
        - predict_df["IW_pred"]
    )

    print("Mean logfO2:",
          round(predict_df["logfO2_pred"].mean(), 3))

    print("Mean ΔIW:",
          round(predict_df["dIW_pred"].mean(), 3), "\n")

    # ==========================================================
    # Diagnostic plots
    # ==========================================================

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    # P–T scatter plot
    axes[0].scatter(
        predict_df["P_pred"],
        predict_df["T_pred"],
        c=predict_df["is_OOD"].map({True: "red", False: "blue"}),
        alpha=0.6
    )

    axes[0].set_xlabel("P_pred")
    axes[0].set_ylabel("T_pred")
    axes[0].set_title("Predicted P–T")

    # IW vs logfO2
    axes[1].scatter(
        predict_df["IW_pred"],
        predict_df["logfO2_pred"],
        c=predict_df["is_OOD"].map({True: "red", False: "blue"}),
        alpha=0.6
    )

    axes[1].plot(
        [predict_df["IW_pred"].min(), predict_df["IW_pred"].max()],
        [predict_df["IW_pred"].min(), predict_df["IW_pred"].max()],
        "k--"
    )

    axes[1].set_xlabel("IW")
    axes[1].set_ylabel("logfO2")

    # ΔIW distribution
    axes[2].hist(predict_df["dIW_pred"], bins=30)
    axes[2].set_xlabel("ΔIW")
    axes[2].set_ylabel("Count")

    plt.tight_layout()
    plt.show()

    # ==========================================================
    # Export results
    # ==========================================================

    output_columns = (
        BASE_FEATURES +
        [
            "P_pred",
            "T_pred",
            "IW_pred",
            "logfO2_pred",
            "dIW_pred",
            "OOD_distance",
            "OOD_level",
            "is_OOD"
        ]
    )

    output_df = predict_df[output_columns]

    output_path = os.path.join(
        RESULT_DIR,
        f"{prediction_name}_oxybarometer_predictions.xlsx"
    )

    output_df.to_excel(output_path, index=False)

    print("Saved to:", output_path)
    print("====================================\n")

    return predict_df
```

```python
# ==========================================================
# SECTION 0 — GLOBAL CONFIGURATION
# ==========================================================

import os
import numpy as np
import torch

print("Initializing FINAL oxybarometer pipeline...")

# ----------------------------------------------------------
# Random Seed
# ----------------------------------------------------------
RANDOM_STATE =42
np.random.seed(RANDOM_STATE)
torch.manual_seed(RANDOM_STATE)

# ----------------------------------------------------------
# Device
# ----------------------------------------------------------
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {DEVICE}")

# ----------------------------------------------------------
# Lithology
# ----------------------------------------------------------
LITHOLOGY_COLUMN = "cluster_label"

LITHOLOGY_1_LABEL = 1
LITHOLOGY_1_NAME  = "Peridotite"

LITHOLOGY_2_LABEL = 2
LITHOLOGY_2_NAME  = "Eclogite"

# ----------------------------------------------------------
# Base Features (composition-derived only)
# ----------------------------------------------------------
BASE_FEATURES = [
    'Si_Grt',
    'Ti_Grt',
    'Al_Grt',
    'Cr_Grt',
    'Fe_Grt',
    'Mg_Grt',
    'Ca_Grt',
    'Na_Grt',
]

print("Base feature count:", len(BASE_FEATURES))

# ----------------------------------------------------------
# Model Selection
# ----------------------------------------------------------
MODEL_TYPE = "tabpfn"   # "tabpfn" | "xgboost" | "rf"

print("Model type:", MODEL_TYPE)

# ----------------------------------------------------------
# Output Directories
# ----------------------------------------------------------
MODEL_DIR  = "models"
RESULT_DIR = "results"
SHAP_DIR   = "shap_results"

os.makedirs(MODEL_DIR, exist_ok=True)
os.makedirs(RESULT_DIR, exist_ok=True)
os.makedirs(SHAP_DIR, exist_ok=True)

print("Output directories ready.")

# ----------------------------------------------------------
# Thermodynamic Function — IW
# (Used everywhere consistently)
# ----------------------------------------------------------
import numpy as np

def calculate_iw(P, T):
    """
    Vectorized IW calculation
    P: GPa (array or scalar)
    T: °C (array or scalar)
    """

    P = np.asarray(P)
    T = np.asarray(T)

    T_kelvin = T + 273.15

    # 相变边界
    x0 = -18.64
    x1 = 0.04359
    x2 = -5.069e-6

    P_transition = x0 + x1 * T_kelvin + x2 * T_kelvin**2

    # ===== fcc/bcc =====
    a = 6.844864 + 1.175691e-1 * P + 1.143873e-3 * P**2
    b = 5.791364e-4 - 2.891434e-4 * P - 2.737171e-7 * P**2
    c = (-7.971469e-5
         + 3.198005e-5 * P
         + 1.059554e-10 * P**3
         + 2.014461e-7 * np.sqrt(P))
    d = (-2.769002e4
         + 5.285977e2 * P
         - 2.919275 * P**2)

    iw_fcc = a + b * T_kelvin + c * T_kelvin * np.log(T_kelvin) + d / T_kelvin

    # ===== hcp =====
    e = 8.463095 - 3.000307e-3 * P + 7.213445e-5 * P**2
    f = 1.148738e-3 - 9.352312e-5 * P + 5.161592e-7 * P**2
    g = (-7.448624e-4
         - 6.329325e-6 * P
         - 1.407339e-10 * P**3
         + 1.830014e-4 * np.sqrt(P))
    h = (-2.782082e4
         + 5.285977e2 * P
         - 8.473231e-1 * P**2)

    iw_hcp = e + f * T_kelvin + g * T_kelvin * np.log(T_kelvin) + h / T_kelvin

    # ===== 关键：逐点选择 =====
    iw = np.where(P <= P_transition, iw_fcc, iw_hcp)

    return iw


print("Thermodynamic IW function loaded.")

print("SECTION 0 completed.\n")
```

```python
# ==========================================================
# MODEL FACTORY
# Switch between TabPFN / XGBoost / RandomForest
# Different parameters for P / T / fO2
# ==========================================================

from tabpfn import TabPFNRegressor
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor


def create_model(model_type, target):

    # ------------------------------------------------------
    # TabPFN
    # ------------------------------------------------------
    if model_type == "tabpfn":

        model = TabPFNRegressor(
            device=DEVICE
        )

    # ------------------------------------------------------
    # XGBoost
    # ------------------------------------------------------
    elif model_type == "xgboost":

        if target == "P":

            model = XGBRegressor(
                n_estimators=3000,
                min_child_weight=2,
                max_depth=8,
                learning_rate=0.01,
                subsample=0.5,
                colsample_bytree=0.9,
                gamma=0,
                random_state=RANDOM_STATE,
                tree_method="hist",
                objective="reg:squarederror",
                n_jobs=1
            )

        elif target == "T":

            model = XGBRegressor(
                n_estimators=3000,
                min_child_weight=3,
                max_depth=14,
                learning_rate=0.01,
                subsample=0.5,
                colsample_bytree=1,
                gamma=0.2,
                random_state=RANDOM_STATE,
                tree_method="hist",
                objective="reg:squarederror",
                n_jobs=1
            )

        elif target == "fO2":

            model = XGBRegressor(
                n_estimators=2000,
                min_child_weight=4,
                max_depth=6,
                learning_rate=0.05,
                subsample=0.9,
                colsample_bytree=0.6,
                gamma=0.2,
                random_state=RANDOM_STATE,
                tree_method="hist",
                objective="reg:squarederror",
                n_jobs=1
            )

        else:
            raise ValueError("Unknown target for XGBoost")

    # ------------------------------------------------------
    # Random Forest
    # ------------------------------------------------------
    elif model_type == "rf":

        if target == "P":

            model = RandomForestRegressor(
                n_estimators=400,
                max_depth=15,
                min_samples_split=2,
                min_samples_leaf=1,
                max_features=0.6,
                random_state=RANDOM_STATE,
                n_jobs=1
            )

        elif target == "T":

            model = RandomForestRegressor(
                n_estimators=2000,
                max_depth=23,
                min_samples_split=2,
                min_samples_leaf=1,
                max_features=0.6,
                random_state=RANDOM_STATE,
                n_jobs=1
            )

        elif target == "fO2":

            model = RandomForestRegressor(
                n_estimators=1500,
                max_depth=13,
                min_samples_split=2,
                min_samples_leaf=1,
                max_features=0.8,
                random_state=RANDOM_STATE,
                n_jobs=1
            )

        else:
            raise ValueError("Unknown target for RF")

    else:
        raise ValueError("Unknown MODEL_TYPE")

    return model


print("Model factory ready.")
```

```python
# ==========================================================
# SECTION 1 — TRAIN P MODEL + OOD DETECTION
# composition → P
# ==========================================================

import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import NearestNeighbors
from tabpfn import TabPFNRegressor
import joblib
import numpy as np
import pandas as pd
import os
import matplotlib as mpl

print("Preparing P model dataset...")

# ----------------------------------------------------------
# Merge Dataset I + II (train / test)
# ----------------------------------------------------------
P_train_df = pd.concat(
    [df_P_T_train, df_fO2_train],
    ignore_index=True
)

P_test_df = pd.concat(
    [df_P_T_test, df_fO2_test],
    ignore_index=True
)

X_train = P_train_df[BASE_FEATURES].copy()
y_train = P_train_df["P"].values

X_test  = P_test_df[BASE_FEATURES].copy()
y_test  = P_test_df["P"].values

stratify_labels = P_train_df[LITHOLOGY_COLUMN].values
test_lithology_labels = P_test_df[LITHOLOGY_COLUMN].values

print("Train size:", len(X_train))
print("Test size :", len(X_test))

# ----------------------------------------------------------
# 10-Fold Stratified CV
# ----------------------------------------------------------
print("\nStarting 10-fold CV...")

N_SPLITS = 10

skf = StratifiedKFold(
    n_splits=N_SPLITS,
    shuffle=True,
    random_state=RANDOM_STATE
)

rmse_list = []
mae_list  = []
r2_list   = []

# ---- per-lithology CV metrics ----
cv_lith_metrics = {
    LITHOLOGY_1_NAME: {"rmse": [], "mae": [], "r2": []},
    LITHOLOGY_2_NAME: {"rmse": [], "mae": [], "r2": []}
}

best_r2 = -np.inf
best_fold_data = None
best_fold_metrics = None

for fold, (tr, val) in enumerate(skf.split(X_train, stratify_labels)):

    X_tr, X_val = X_train.iloc[tr], X_train.iloc[val]
    y_tr, y_val = y_train[tr], y_train[val]
    lith_val = stratify_labels[val]

    model = create_model(MODEL_TYPE, 'P')
    model.fit(X_tr.values, y_tr)

    y_val_pred = model.predict(X_val.values)

    rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
    mae  = mean_absolute_error(y_val, y_val_pred)
    r2   = r2_score(y_val, y_val_pred)

    rmse_list.append(rmse)
    mae_list.append(mae)
    r2_list.append(r2)

    print(f"Fold {fold+1}: RMSE={rmse:.4f}, MAE={mae:.4f}, R2={r2:.4f}")

    # ---- per lithology in this fold ----
    for lith_label, lith_name in [
        (LITHOLOGY_1_LABEL, LITHOLOGY_1_NAME),
        (LITHOLOGY_2_LABEL, LITHOLOGY_2_NAME)
    ]:
        mask = lith_val == lith_label
        if np.sum(mask) >= 2:
            cv_lith_metrics[lith_name]["rmse"].append(
                np.sqrt(mean_squared_error(y_val[mask], y_val_pred[mask]))
            )
            cv_lith_metrics[lith_name]["mae"].append(
                mean_absolute_error(y_val[mask], y_val_pred[mask])
            )
            cv_lith_metrics[lith_name]["r2"].append(
                r2_score(y_val[mask], y_val_pred[mask])
            )

    if r2 > best_r2:
        best_r2 = r2
        best_fold_data = (y_val, y_val_pred, lith_val)
        best_fold_metrics = (rmse, mae, r2)

# ----------------------------------------------------------
# CV Summary
# ----------------------------------------------------------
print("\nCV Results (Mean ± Std):")
print(f"RMSE: {np.mean(rmse_list):.4f} ± {np.std(rmse_list):.4f}")
print(f"MAE : {np.mean(mae_list):.4f} ± {np.std(mae_list):.4f}")
print(f"R2  : {np.mean(r2_list):.4f} ± {np.std(r2_list):.4f}")

print("\nCV Results by Lithology (Mean ± Std):")
for lith_name in [LITHOLOGY_1_NAME, LITHOLOGY_2_NAME]:
    print(f"\n{lith_name}:")
    print(f"  RMSE: {np.mean(cv_lith_metrics[lith_name]['rmse']):.4f} ± {np.std(cv_lith_metrics[lith_name]['rmse']):.4f}")
    print(f"  MAE : {np.mean(cv_lith_metrics[lith_name]['mae']):.4f} ± {np.std(cv_lith_metrics[lith_name]['mae']):.4f}")
    print(f"  R2  : {np.mean(cv_lith_metrics[lith_name]['r2']):.4f} ± {np.std(cv_lith_metrics[lith_name]['r2']):.4f}")

# ----------------------------------------------------------
# Train final model on full training set
# ----------------------------------------------------------
print("\nTraining final P_model...")

P_model = create_model(MODEL_TYPE, 'P')
P_model.fit(X_train.values, y_train)

# ----------------------------------------------------------
# OOD Detection on Test Set
# ----------------------------------------------------------
print("\nRunning OOD detection on test set...")

# Standardization based on training set only
ood_scaler = StandardScaler()
X_train_scaled = ood_scaler.fit_transform(X_train)
X_test_scaled = ood_scaler.transform(X_test)

# 1-NN distance from each test sample to training distribution
nn_test = NearestNeighbors(n_neighbors=1)
nn_test.fit(X_train_scaled)

test_distances, _ = nn_test.kneighbors(X_test_scaled)
test_distances = test_distances.flatten()

# Internal train nearest-neighbor distances
# n_neighbors=2 because the nearest one is itself
nn_train = NearestNeighbors(n_neighbors=2)
nn_train.fit(X_train_scaled)

train_distances, _ = nn_train.kneighbors(X_train_scaled)
train_nn_dist = train_distances[:, 1]

# OOD threshold: 99th percentile of train NN distance
threshold_ood = np.percentile(train_nn_dist, 99)

# OOD mask for test samples
test_ood_mask = test_distances > threshold_ood

# Store OOD info in test dataframe
P_test_df = P_test_df.copy()
P_test_df["OOD_distance"] = test_distances
P_test_df["OOD_level"] = "In-Distribution"
P_test_df.loc[test_ood_mask, "OOD_level"] = "OOD"
P_test_df["is_OOD"] = test_ood_mask

print(f"OOD threshold (99th percentile of train NN distance): {threshold_ood:.4f}")
print(f"Test OOD ratio: {np.mean(test_ood_mask) * 100:.2f}%")
print(f"Test OOD count: {np.sum(test_ood_mask)} / {len(test_ood_mask)}")

# PCA projection for visualization
pca = PCA(n_components=2, random_state=RANDOM_STATE)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# ----------------------------------------------------------
# Test evaluation
# ----------------------------------------------------------
y_test_pred = P_model.predict(X_test.values)

test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
test_mae  = mean_absolute_error(y_test, y_test_pred)
test_r2   = r2_score(y_test, y_test_pred)

print("\nTest Results:")
print(f"RMSE: {test_rmse:.4f}")
print(f"MAE : {test_mae:.4f}")
print(f"R2  : {test_r2:.4f}")

# ---- overall test results split by OOD ----
if np.sum(~test_ood_mask) > 0:
    print("\nIn-Distribution Test Samples:")
    print(f"  N   : {np.sum(~test_ood_mask)}")
    print(f"  RMSE: {np.sqrt(mean_squared_error(y_test[~test_ood_mask], y_test_pred[~test_ood_mask])):.4f}")
    print(f"  MAE : {mean_absolute_error(y_test[~test_ood_mask], y_test_pred[~test_ood_mask]):.4f}")
    if np.sum(~test_ood_mask) >= 2:
        print(f"  R2  : {r2_score(y_test[~test_ood_mask], y_test_pred[~test_ood_mask]):.4f}")

if np.sum(test_ood_mask) > 0:
    print("\nOOD Test Samples:")
    print(f"  N   : {np.sum(test_ood_mask)}")
    print(f"  RMSE: {np.sqrt(mean_squared_error(y_test[test_ood_mask], y_test_pred[test_ood_mask])):.4f}")
    print(f"  MAE : {mean_absolute_error(y_test[test_ood_mask], y_test_pred[test_ood_mask]):.4f}")
    if np.sum(test_ood_mask) >= 2:
        print(f"  R2  : {r2_score(y_test[test_ood_mask], y_test_pred[test_ood_mask]):.4f}")

# ---- test by lithology ----
print("\nTest Results by Lithology:")
for lith_label, lith_name in [
    (LITHOLOGY_1_LABEL, LITHOLOGY_1_NAME),
    (LITHOLOGY_2_LABEL, LITHOLOGY_2_NAME)
]:
    mask = test_lithology_labels == lith_label
    y_true_l = y_test[mask]
    y_pred_l = y_test_pred[mask]
    residual_l = y_true_l - y_pred_l
    abs_error_l = np.abs(residual_l)

    print(f"\n{lith_name}:")
    print(f"  N   : {np.sum(mask)}")
    print(f"  RMSE: {np.sqrt(mean_squared_error(y_true_l, y_pred_l)):.4f}")
    print(f"  MAE : {mean_absolute_error(y_true_l, y_pred_l):.4f}")
    if np.sum(mask) >= 2:
        print(f"  R2  : {r2_score(y_true_l, y_pred_l):.4f}")
    print(f"  Residual mean ± std : {np.mean(residual_l):.4f} ± {np.std(residual_l):.4f}")
    print(f"  Abs error mean ± std: {np.mean(abs_error_l):.4f} ± {np.std(abs_error_l):.4f}")

    lith_ood_n = np.sum(test_ood_mask[mask])
    print(f"  OOD count: {lith_ood_n} / {np.sum(mask)}")

# ----------------------------------------------------------
# Visualization
# ----------------------------------------------------------
mpl.rcParams["font.family"] = "Arial"
mpl.rcParams["font.size"] = 11

color_1 = "#3D5A80"
color_2 = "#D08C60"
ood_color = "#C1121F"
line_color = "gray"

# ----------------------------------------------------------
# Figure 1: Best CV + Test Set with OOD highlight
# ----------------------------------------------------------
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# ----------------------------------------------------------
# Best CV fold
# ----------------------------------------------------------
y_true_best, y_pred_best, lith_best = best_fold_data
rmse_best, mae_best, r2_best = best_fold_metrics

mask_1 = lith_best == LITHOLOGY_1_LABEL
mask_2 = lith_best == LITHOLOGY_2_LABEL

axes[0].scatter(
    y_true_best[mask_1],
    y_pred_best[mask_1],
    s=80,
    color=color_1,
    alpha=0.7,
    edgecolors="none",
    label=LITHOLOGY_1_NAME
)

axes[0].scatter(
    y_true_best[mask_2],
    y_pred_best[mask_2],
    s=80,
    color=color_2,
    alpha=0.7,
    edgecolors="none",
    label=LITHOLOGY_2_NAME
)

min_val = min(y_true_best.min(), y_pred_best.min())
max_val = max(y_true_best.max(), y_pred_best.max())

axes[0].plot(
    [min_val, max_val],
    [min_val, max_val],
    linestyle="--",
    color=line_color,
    linewidth=1.5
)

axes[0].set_title("Best CV Fold")
axes[0].set_xlabel("True P")
axes[0].set_ylabel("Predicted P")

axes[0].text(
    0.05, 0.95,
    f"RMSE = {rmse_best:.3f}\nMAE = {mae_best:.3f}\nR² = {r2_best:.3f}",
    transform=axes[0].transAxes,
    verticalalignment="top",
    bbox=dict(facecolor="white", edgecolor="none", alpha=0.8)
)

axes[0].legend(frameon=False, loc="lower right")
axes[0].set_aspect("equal", adjustable="box")
axes[0].grid(False)

# ----------------------------------------------------------
# Test set with OOD highlight
# ----------------------------------------------------------
mask_1_test = test_lithology_labels == LITHOLOGY_1_LABEL
mask_2_test = test_lithology_labels == LITHOLOGY_2_LABEL

# In-distribution points
axes[1].scatter(
    y_test[mask_1_test & ~test_ood_mask],
    y_test_pred[mask_1_test & ~test_ood_mask],
    s=80,
    color=color_1,
    alpha=0.7,
    edgecolors="none",
    label=f"{LITHOLOGY_1_NAME} (ID)"
)

axes[1].scatter(
    y_test[mask_2_test & ~test_ood_mask],
    y_test_pred[mask_2_test & ~test_ood_mask],
    s=80,
    color=color_2,
    alpha=0.7,
    edgecolors="none",
    label=f"{LITHOLOGY_2_NAME} (ID)"
)

# OOD points highlighted
axes[1].scatter(
    y_test[test_ood_mask],
    y_test_pred[test_ood_mask],
    s=110,
    facecolors=ood_color,
    edgecolors="black",
    linewidths=0.8,
    alpha=0.95,
    label="OOD"
)

min_val = min(y_test.min(), y_test_pred.min())
max_val = max(y_test.max(), y_test_pred.max())

axes[1].plot(
    [min_val, max_val],
    [min_val, max_val],
    linestyle="--",
    color=line_color,
    linewidth=1.5
)

axes[1].set_title("Test Set")
axes[1].set_xlabel("True P")
axes[1].set_ylabel("Predicted P")

axes[1].text(
    0.05, 0.95,
    f"RMSE = {test_rmse:.3f}\nMAE = {test_mae:.3f}\nR² = {test_r2:.3f}\nOOD = {np.sum(test_ood_mask)}",
    transform=axes[1].transAxes,
    verticalalignment="top",
    bbox=dict(facecolor="white", edgecolor="none", alpha=0.8)
)

axes[1].legend(frameon=False, loc="lower right")
axes[1].set_aspect("equal", adjustable="box")
axes[1].grid(False)

for ax in axes:
    for spine in ax.spines.values():
        spine.set_linewidth(2)

plt.tight_layout()

save_path_png = os.path.join(IMAGE_PATH, "Fig_a_P_model.png")
save_path_pdf = os.path.join(IMAGE_PATH, "Fig_a_P_model.pdf")

plt.savefig(save_path_png, dpi=600)
plt.savefig(save_path_pdf)
plt.show()

# ----------------------------------------------------------
# Figure 2: PCA feature-space OOD visualization
# ----------------------------------------------------------
plt.figure(figsize=(8, 6))

plt.scatter(
    X_train_pca[:, 0],
    X_train_pca[:, 1],
    alpha=0.25,
    color="gray",
    s=45,
    label="Train"
)

plt.scatter(
    X_test_pca[~test_ood_mask, 0],
    X_test_pca[~test_ood_mask, 1],
    alpha=0.75,
    color="#4C78A8",
    s=70,
    label="Test (ID)"
)

if np.sum(test_ood_mask) > 0:
    plt.scatter(
        X_test_pca[test_ood_mask, 0],
        X_test_pca[test_ood_mask, 1],
        alpha=0.95,
        color=ood_color,
        edgecolors="black",
        linewidths=0.8,
        s=90,
        label="Test (OOD)"
    )

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("P Model Feature Space PCA Projection")
plt.legend(frameon=False)
plt.tight_layout()

save_path_pca_png = os.path.join(IMAGE_PATH, "Fig_a_P_model_OOD_PCA.png")
save_path_pca_pdf = os.path.join(IMAGE_PATH, "Fig_a_P_model_OOD_PCA.pdf")

plt.savefig(save_path_pca_png, dpi=600)
plt.savefig(save_path_pca_pdf)
plt.show()

# ----------------------------------------------------------
# Save model
# ----------------------------------------------------------
joblib.dump(P_model, os.path.join(MODEL_PATH, f"P_model_{MODEL_TYPE}.pkl"))
print("\nP model saved.")

# ----------------------------------------------------------
# Save test predictions with OOD information
# ----------------------------------------------------------
P_test_result = pd.DataFrame({
    "P_true": y_test,
    "P_pred": y_test_pred,
    "lithology": test_lithology_labels,
    "OOD_distance": test_distances,
    "OOD_level": np.where(test_ood_mask, "OOD", "In-Distribution"),
    "is_OOD": test_ood_mask
})

P_test_result["residual"] = (
    P_test_result["P_true"] - P_test_result["P_pred"]
)
P_test_result["abs_error"] = np.abs(
    P_test_result["residual"]
)

P_test_result.to_excel(
    os.path.join(RESULT_DIR, "P_test_predictions.xlsx"),
    index=False
)

# ----------------------------------------------------------
# Feature importance
# ----------------------------------------------------------
if MODEL_TYPE in ["xgboost", "rf"]:

    print("\nComputing feature importance...")

    importance = P_model.feature_importances_

    importance_df = pd.DataFrame({
        "feature": BASE_FEATURES,
        "importance": importance
    }).sort_values(
        "importance",
        ascending=False
    )

    print("\nFeature Importance:")
    print(importance_df)

    plt.figure(figsize=(6, 4))

    plt.barh(
        importance_df["feature"],
        importance_df["importance"]
    )

    plt.gca().invert_yaxis()

    plt.xlabel("Importance")
    plt.title(f"{MODEL_TYPE.upper()} Feature Importance")

    plt.tight_layout()

    plt.savefig(
        os.path.join(
            IMAGE_PATH,
            f"{MODEL_TYPE}_feature_importance.png"
        ),
        dpi=300
    )

    plt.show()

    importance_df.to_excel(
        os.path.join(
            RESULT_DIR,
            f"{MODEL_TYPE}_feature_importance.xlsx"
        ),
        index=False
    )

    print("Feature importance saved.\n")

print("P test predictions with OOD info saved.")
```

```python
# ==========================================================
# SECTION 1 — TRAIN P MODEL
# composition → P
# ==========================================================

import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tabpfn import TabPFNRegressor
import joblib
import numpy as np
import pandas as pd
import os

print("Preparing P model dataset...")

# ----------------------------------------------------------
# Merge Dataset I + II (train)
# ----------------------------------------------------------
P_train_df = pd.concat(
    [df_P_T_train, df_fO2_train],
    ignore_index=True
)

P_test_df = pd.concat(
    [df_P_T_test, df_fO2_test],
    ignore_index=True
)

X_train = P_train_df[BASE_FEATURES].copy()
y_train = P_train_df["P"].values

X_test  = P_test_df[BASE_FEATURES].copy()
y_test  = P_test_df["P"].values

stratify_labels = P_train_df[LITHOLOGY_COLUMN].values
test_lithology_labels = P_test_df[LITHOLOGY_COLUMN].values

print("Train size:", len(X_train))
print("Test size :", len(X_test))

# ----------------------------------------------------------
# 10-Fold Stratified CV
# ----------------------------------------------------------
print("\nStarting 10-fold CV...")

N_SPLITS = 10

skf = StratifiedKFold(
    n_splits=N_SPLITS,
    shuffle=True,
    random_state=RANDOM_STATE
)

rmse_list = []
mae_list  = []
r2_list   = []

# ---- per-lithology CV metrics ----
cv_lith_metrics = {
    LITHOLOGY_1_NAME: {"rmse": [], "mae": [], "r2": []},
    LITHOLOGY_2_NAME: {"rmse": [], "mae": [], "r2": []}
}

best_r2 = -np.inf
best_fold_data = None
best_fold_metrics = None

for fold, (tr, val) in enumerate(skf.split(X_train, stratify_labels)):

    X_tr, X_val = X_train.iloc[tr], X_train.iloc[val]
    y_tr, y_val = y_train[tr], y_train[val]
    lith_val = stratify_labels[val]

    model = create_model(MODEL_TYPE, 'P')
    model.fit(X_tr.values, y_tr)

    y_val_pred = model.predict(X_val.values)

    rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
    mae  = mean_absolute_error(y_val, y_val_pred)
    r2   = r2_score(y_val, y_val_pred)

    rmse_list.append(rmse)
    mae_list.append(mae)
    r2_list.append(r2)

    print(f"Fold {fold+1}: RMSE={rmse:.4f}, MAE={mae:.4f}, R2={r2:.4f}")

    # ---- per lithology in this fold ----
    for lith_label, lith_name in [
        (LITHOLOGY_1_LABEL, LITHOLOGY_1_NAME),
        (LITHOLOGY_2_LABEL, LITHOLOGY_2_NAME)
    ]:
        mask = lith_val == lith_label
        if np.sum(mask) >= 2:
            cv_lith_metrics[lith_name]["rmse"].append(
                np.sqrt(mean_squared_error(y_val[mask], y_val_pred[mask]))
            )
            cv_lith_metrics[lith_name]["mae"].append(
                mean_absolute_error(y_val[mask], y_val_pred[mask])
            )
            cv_lith_metrics[lith_name]["r2"].append(
                r2_score(y_val[mask], y_val_pred[mask])
            )

    if r2 > best_r2:
        best_r2 = r2
        best_fold_data = (y_val, y_val_pred, lith_val)
        best_fold_metrics = (rmse, mae, r2)

# ----------------------------------------------------------
# CV Summary
# ----------------------------------------------------------
print("\nCV Results (Mean ± Std):")
print(f"RMSE: {np.mean(rmse_list):.4f} ± {np.std(rmse_list):.4f}")
print(f"MAE : {np.mean(mae_list):.4f} ± {np.std(mae_list):.4f}")
print(f"R2  : {np.mean(r2_list):.4f} ± {np.std(r2_list):.4f}")

print("\nCV Results by Lithology (Mean ± Std):")
for lith_name in [LITHOLOGY_1_NAME, LITHOLOGY_2_NAME]:
    print(f"\n{lith_name}:")
    print(f"  RMSE: {np.mean(cv_lith_metrics[lith_name]['rmse']):.4f} ± {np.std(cv_lith_metrics[lith_name]['rmse']):.4f}")
    print(f"  MAE : {np.mean(cv_lith_metrics[lith_name]['mae']):.4f} ± {np.std(cv_lith_metrics[lith_name]['mae']):.4f}")
    print(f"  R2  : {np.mean(cv_lith_metrics[lith_name]['r2']):.4f} ± {np.std(cv_lith_metrics[lith_name]['r2']):.4f}")

# ----------------------------------------------------------
# Train final model on full training set
# ----------------------------------------------------------
print("\nTraining final P_model...")

P_model = create_model(MODEL_TYPE, 'P')
P_model.fit(X_train.values, y_train)

# ----------------------------------------------------------
# Test evaluation
# ----------------------------------------------------------
y_test_pred = P_model.predict(X_test.values)

test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
test_mae  = mean_absolute_error(y_test, y_test_pred)
test_r2   = r2_score(y_test, y_test_pred)

print("\nTest Results:")
print(f"RMSE: {test_rmse:.4f}")
print(f"MAE : {test_mae:.4f}")
print(f"R2  : {test_r2:.4f}")

# ---- test by lithology ----
print("\nTest Results by Lithology:")
for lith_label, lith_name in [
    (LITHOLOGY_1_LABEL, LITHOLOGY_1_NAME),
    (LITHOLOGY_2_LABEL, LITHOLOGY_2_NAME)
]:
    mask = test_lithology_labels == lith_label
    y_true_l = y_test[mask]
    y_pred_l = y_test_pred[mask]
    residual_l = y_true_l - y_pred_l
    abs_error_l = np.abs(residual_l)

    print(f"\n{lith_name}:")
    print(f"  N   : {np.sum(mask)}")
    print(f"  RMSE: {np.sqrt(mean_squared_error(y_true_l, y_pred_l)):.4f}")
    print(f"  MAE : {mean_absolute_error(y_true_l, y_pred_l):.4f}")
    print(f"  R2  : {r2_score(y_true_l, y_pred_l):.4f}")
    print(f"  Residual mean ± std : {np.mean(residual_l):.4f} ± {np.std(residual_l):.4f}")
    print(f"  Abs error mean ± std: {np.mean(abs_error_l):.4f} ± {np.std(abs_error_l):.4f}")

# ----------------------------------------------------------
# Visualization
# ----------------------------------------------------------

import matplotlib as mpl

mpl.rcParams["font.family"] = "Arial"
mpl.rcParams["font.size"] = 11

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

color_1 = "#3D5A80"
color_2 = "#D08C60"
line_color = "gray"

# ----------------------------------------------------------
# Best CV fold
# ----------------------------------------------------------

y_true_best, y_pred_best, lith_best = best_fold_data
rmse_best, mae_best, r2_best = best_fold_metrics

mask_1 = lith_best == LITHOLOGY_1_LABEL
mask_2 = lith_best == LITHOLOGY_2_LABEL

axes[0].scatter(
    y_true_best[mask_1],
    y_pred_best[mask_1],
    s=80,
    color=color_1,
    alpha=0.7,
    edgecolors="none",
    label=LITHOLOGY_1_NAME
)

axes[0].scatter(
    y_true_best[mask_2],
    y_pred_best[mask_2],
    s=80,
    color=color_2,
    alpha=0.7,
    edgecolors="none",
    label=LITHOLOGY_2_NAME
)

min_val = min(y_true_best.min(), y_pred_best.min())
max_val = max(y_true_best.max(), y_pred_best.max())

axes[0].plot(
    [min_val, max_val],
    [min_val, max_val],
    linestyle="--",
    color=line_color,
    linewidth=1.5
)

axes[0].set_title("Best CV Fold")
axes[0].set_xlabel("True P")
axes[0].set_ylabel("Predicted P")

axes[0].text(
    0.05, 0.95,
    f"RMSE = {rmse_best:.3f}\nMAE = {mae_best:.3f}\nR² = {r2_best:.3f}",
    transform=axes[0].transAxes,
    verticalalignment="top",
    bbox=dict(facecolor="white", edgecolor="none", alpha=0.8)
)

axes[0].legend(frameon=False, loc="lower right")
axes[0].set_aspect("equal", adjustable="box")
axes[0].grid(False)

# ----------------------------------------------------------
# Test set
# ----------------------------------------------------------

mask_1_test = test_lithology_labels == LITHOLOGY_1_LABEL
mask_2_test = test_lithology_labels == LITHOLOGY_2_LABEL

axes[1].scatter(
    y_test[mask_1_test],
    y_test_pred[mask_1_test],
    s=80,
    color=color_1,
    alpha=0.7,
    edgecolors="none",
    label=LITHOLOGY_1_NAME
)

axes[1].scatter(
    y_test[mask_2_test],
    y_test_pred[mask_2_test],
    s=80,
    color=color_2,
    alpha=0.7,
    edgecolors="none",
    label=LITHOLOGY_2_NAME
)

min_val = min(y_test.min(), y_test_pred.min())
max_val = max(y_test.max(), y_test_pred.max())

axes[1].plot(
    [min_val, max_val],
    [min_val, max_val],
    linestyle="--",
    color=line_color,
    linewidth=1.5
)

axes[1].set_title("Test Set")
axes[1].set_xlabel("True P")
axes[1].set_ylabel("Predicted P")

axes[1].text(
    0.05, 0.95,
    f"RMSE = {test_rmse:.3f}\nMAE = {test_mae:.3f}\nR² = {test_r2:.3f}",
    transform=axes[1].transAxes,
    verticalalignment="top",
    bbox=dict(facecolor="white", edgecolor="none", alpha=0.8)
)

axes[1].legend(frameon=False, loc="lower right")
axes[1].set_aspect("equal", adjustable="box")
axes[1].grid(False)

for ax in axes:
    for spine in ax.spines.values():
        spine.set_linewidth(2)

plt.tight_layout()

save_path_png = os.path.join(IMAGE_PATH, "Fig_a_P_model.png")
save_path_pdf = os.path.join(IMAGE_PATH, "Fig_a_P_model.pdf")

plt.savefig(save_path_png, dpi=600)
plt.savefig(save_path_pdf)

plt.show()

joblib.dump(P_model, os.path.join(MODEL_PATH, f"P_model_{MODEL_TYPE}.pkl"))
print("\nP model saved.")

P_test_result = pd.DataFrame({
    "P_true": y_test,
    "P_pred": y_test_pred,
    "lithology": test_lithology_labels
})

P_test_result["residual"] = (
    P_test_result["P_true"] - P_test_result["P_pred"]
)
P_test_result["abs_error"] = np.abs(
    P_test_result["residual"]
)

P_test_result.to_excel(
    os.path.join(RESULT_DIR, "P_test_predictions.xlsx"),
    index=False
)

if MODEL_TYPE in ["xgboost", "rf"]:

    print("\nComputing feature importance...")

    importance = P_model.feature_importances_

    importance_df = pd.DataFrame({
        "feature": BASE_FEATURES,
        "importance": importance
    }).sort_values(
        "importance",
        ascending=False
    )

    print("\nFeature Importance:")
    print(importance_df)

    plt.figure(figsize=(6, 4))

    plt.barh(
        importance_df["feature"],
        importance_df["importance"]
    )

    plt.gca().invert_yaxis()

    plt.xlabel("Importance")
    plt.title(f"{MODEL_TYPE.upper()} Feature Importance")

    plt.tight_layout()

    plt.savefig(
        os.path.join(
            IMAGE_PATH,
            f"{MODEL_TYPE}_feature_importance.png"
        ),
        dpi=300
    )

    plt.show()

    importance_df.to_excel(
        os.path.join(
            RESULT_DIR,
            f"{MODEL_TYPE}_feature_importance.xlsx"
        ),
        index=False
    )

    print("Feature importance saved.\n")

print("P test predictions saved.")
```

```python
# ==========================================================
# SECTION 2 — TRAIN T MODEL
# composition → T
# ==========================================================

import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tabpfn import TabPFNRegressor
import joblib
import numpy as np
import pandas as pd
import os

print("Preparing T model dataset...")

T_train_df = pd.concat(
    [df_P_T_train, df_fO2_train],
    ignore_index=True
)

T_test_df = pd.concat(
    [df_P_T_test, df_fO2_test],
    ignore_index=True
)

X_train = T_train_df[BASE_FEATURES].copy()
y_train = T_train_df["T"].values

X_test  = T_test_df[BASE_FEATURES].copy()
y_test  = T_test_df["T"].values

stratify_labels = T_train_df[LITHOLOGY_COLUMN].values
test_lithology_labels = T_test_df[LITHOLOGY_COLUMN].values

print("Train size:", len(X_train))
print("Test size :", len(X_test))

print("\nStarting 10-fold CV...")

N_SPLITS = 10

skf = StratifiedKFold(
    n_splits=N_SPLITS,
    shuffle=True,
    random_state=RANDOM_STATE
)

rmse_list = []
mae_list  = []
r2_list   = []

cv_lith_metrics = {
    LITHOLOGY_1_NAME: {"rmse": [], "mae": [], "r2": []},
    LITHOLOGY_2_NAME: {"rmse": [], "mae": [], "r2": []}
}

best_r2 = -np.inf
best_fold_data = None
best_fold_metrics = None

for fold, (tr, val) in enumerate(skf.split(X_train, stratify_labels)):

    X_tr, X_val = X_train.iloc[tr], X_train.iloc[val]
    y_tr, y_val = y_train[tr], y_train[val]
    lith_val = stratify_labels[val]

    model = create_model(MODEL_TYPE, 'T')
    model.fit(X_tr.values, y_tr)

    y_val_pred = model.predict(X_val.values)

    rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
    mae  = mean_absolute_error(y_val, y_val_pred)
    r2   = r2_score(y_val, y_val_pred)

    rmse_list.append(rmse)
    mae_list.append(mae)
    r2_list.append(r2)

    print(f"Fold {fold+1}: RMSE={rmse:.4f}, MAE={mae:.4f}, R2={r2:.4f}")

    for lith_label, lith_name in [
        (LITHOLOGY_1_LABEL, LITHOLOGY_1_NAME),
        (LITHOLOGY_2_LABEL, LITHOLOGY_2_NAME)
    ]:
        mask = lith_val == lith_label
        if np.sum(mask) >= 2:
            cv_lith_metrics[lith_name]["rmse"].append(
                np.sqrt(mean_squared_error(y_val[mask], y_val_pred[mask]))
            )
            cv_lith_metrics[lith_name]["mae"].append(
                mean_absolute_error(y_val[mask], y_val_pred[mask])
            )
            cv_lith_metrics[lith_name]["r2"].append(
                r2_score(y_val[mask], y_val_pred[mask])
            )

    if r2 > best_r2:
        best_r2 = r2
        best_fold_data = (y_val, y_val_pred, lith_val)
        best_fold_metrics = (rmse, mae, r2)

print("\nCV Results (Mean ± Std):")
print(f"RMSE: {np.mean(rmse_list):.4f} ± {np.std(rmse_list):.4f}")
print(f"MAE : {np.mean(mae_list):.4f} ± {np.std(mae_list):.4f}")
print(f"R2  : {np.mean(r2_list):.4f} ± {np.std(r2_list):.4f}")

print("\nCV Results by Lithology (Mean ± Std):")
for lith_name in [LITHOLOGY_1_NAME, LITHOLOGY_2_NAME]:
    print(f"\n{lith_name}:")
    print(f"  RMSE: {np.mean(cv_lith_metrics[lith_name]['rmse']):.4f} ± {np.std(cv_lith_metrics[lith_name]['rmse']):.4f}")
    print(f"  MAE : {np.mean(cv_lith_metrics[lith_name]['mae']):.4f} ± {np.std(cv_lith_metrics[lith_name]['mae']):.4f}")
    print(f"  R2  : {np.mean(cv_lith_metrics[lith_name]['r2']):.4f} ± {np.std(cv_lith_metrics[lith_name]['r2']):.4f}")

print("\nTraining final T_model...")

T_model = create_model(MODEL_TYPE, 'T')
T_model.fit(X_train.values, y_train)

y_test_pred = T_model.predict(X_test.values)

test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
test_mae  = mean_absolute_error(y_test, y_test_pred)
test_r2   = r2_score(y_test, y_test_pred)

print("\nTest Results:")
print(f"RMSE: {test_rmse:.4f}")
print(f"MAE : {test_mae:.4f}")
print(f"R2  : {test_r2:.4f}")

print("\nTest Results by Lithology:")
for lith_label, lith_name in [
    (LITHOLOGY_1_LABEL, LITHOLOGY_1_NAME),
    (LITHOLOGY_2_LABEL, LITHOLOGY_2_NAME)
]:
    mask = test_lithology_labels == lith_label
    y_true_l = y_test[mask]
    y_pred_l = y_test_pred[mask]
    residual_l = y_true_l - y_pred_l
    abs_error_l = np.abs(residual_l)

    print(f"\n{lith_name}:")
    print(f"  N   : {np.sum(mask)}")
    print(f"  RMSE: {np.sqrt(mean_squared_error(y_true_l, y_pred_l)):.4f}")
    print(f"  MAE : {mean_absolute_error(y_true_l, y_pred_l):.4f}")
    print(f"  R2  : {r2_score(y_true_l, y_pred_l):.4f}")
    print(f"  Residual mean ± std : {np.mean(residual_l):.4f} ± {np.std(residual_l):.4f}")
    print(f"  Abs error mean ± std: {np.mean(abs_error_l):.4f} ± {np.std(abs_error_l):.4f}")

import matplotlib as mpl

mpl.rcParams["font.family"] = "Arial"
mpl.rcParams["font.size"] = 11

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

color_1 = "#3D5A80"
color_2 = "#D08C60"
line_color = "gray"

y_true_best, y_pred_best, lith_best = best_fold_data
rmse_best, mae_best, r2_best = best_fold_metrics

mask_1 = lith_best == LITHOLOGY_1_LABEL
mask_2 = lith_best == LITHOLOGY_2_LABEL

axes[0].scatter(
    y_true_best[mask_1],
    y_pred_best[mask_1],
    s=80,
    color=color_1,
    alpha=0.7,
    edgecolors="none",
    label=LITHOLOGY_1_NAME
)

axes[0].scatter(
    y_true_best[mask_2],
    y_pred_best[mask_2],
    s=80,
    color=color_2,
    alpha=0.7,
    edgecolors="none",
    label=LITHOLOGY_2_NAME
)

min_val = min(y_true_best.min(), y_pred_best.min())
max_val = max(y_true_best.max(), y_pred_best.max())

axes[0].plot(
    [min_val, max_val],
    [min_val, max_val],
    linestyle="--",
    color=line_color,
    linewidth=1.5
)

axes[0].set_title("Best CV Fold")
axes[0].set_xlabel("True T")
axes[0].set_ylabel("Predicted T")

axes[0].text(
    0.05, 0.95,
    f"RMSE = {rmse_best:.3f}\nMAE = {mae_best:.3f}\nR² = {r2_best:.3f}",
    transform=axes[0].transAxes,
    verticalalignment="top",
    bbox=dict(facecolor="white", edgecolor="none", alpha=0.8)
)

axes[0].legend(frameon=False, loc="lower right")
axes[0].set_aspect("equal", adjustable="box")
axes[0].grid(False)

mask_1_test = test_lithology_labels == LITHOLOGY_1_LABEL
mask_2_test = test_lithology_labels == LITHOLOGY_2_LABEL

axes[1].scatter(
    y_test[mask_1_test],
    y_test_pred[mask_1_test],
    s=80,
    color=color_1,
    alpha=0.7,
    edgecolors="none",
    label=LITHOLOGY_1_NAME
)

axes[1].scatter(
    y_test[mask_2_test],
    y_test_pred[mask_2_test],
    s=80,
    color=color_2,
    alpha=0.7,
    edgecolors="none",
    label=LITHOLOGY_2_NAME
)

min_val = min(y_test.min(), y_test_pred.min())
max_val = max(y_test.max(), y_test_pred.max())

axes[1].plot(
    [min_val, max_val],
    [min_val, max_val],
    linestyle="--",
    color=line_color,
    linewidth=1.5
)

axes[1].set_title("Test Set")
axes[1].set_xlabel("True T")
axes[1].set_ylabel("Predicted T")

axes[1].text(
    0.05, 0.95,
    f"RMSE = {test_rmse:.3f}\nMAE = {test_mae:.3f}\nR² = {test_r2:.3f}",
    transform=axes[1].transAxes,
    verticalalignment="top",
    bbox=dict(facecolor="white", edgecolor="none", alpha=0.8)
)

axes[1].legend(frameon=False, loc="lower right")
axes[1].set_aspect("equal", adjustable="box")
axes[1].grid(False)

for ax in axes:
    for spine in ax.spines.values():
        spine.set_linewidth(2)

plt.tight_layout()


save_path_png = os.path.join(IMAGE_PATH, "Fig_b_T_model.png")
save_path_pdf = os.path.join(IMAGE_PATH, "Fig_b_T_model.pdf")

plt.savefig(save_path_png, dpi=600)
plt.savefig(save_path_pdf)

plt.show()


joblib.dump(T_model, os.path.join(MODEL_PATH, f"T_model_{MODEL_TYPE}.pkl"))
print("\nT model saved.")

T_test_result = pd.DataFrame({
    "T_true": y_test,
    "T_pred": y_test_pred,
    "lithology": test_lithology_labels
})

T_test_result["residual"] = (
    T_test_result["T_true"] - T_test_result["T_pred"]
)
T_test_result["abs_error"] = np.abs(
    T_test_result["residual"]
)

T_test_result.to_excel(
    os.path.join(RESULT_DIR, "T_test_predictions.xlsx"),
    index=False
)

if MODEL_TYPE in ["xgboost", "rf"]:

    print("\nComputing feature importance...")

    importance = T_model.feature_importances_

    importance_df = pd.DataFrame({
        "feature": BASE_FEATURES,
        "importance": importance
    }).sort_values(
        "importance",
        ascending=False
    )

    print("\nFeature Importance:")
    print(importance_df)

    plt.figure(figsize=(6, 4))

    plt.barh(
        importance_df["feature"],
        importance_df["importance"]
    )

    plt.gca().invert_yaxis()

    plt.xlabel("Importance")
    plt.title(f"{MODEL_TYPE.upper()} Feature Importance")

    plt.tight_layout()

    plt.savefig(
        os.path.join(
            IMAGE_PATH,
            f"{MODEL_TYPE}_feature_importance.png"
        ),
        dpi=300
    )

    plt.show()

    importance_df.to_excel(
        os.path.join(
            RESULT_DIR,
            f"{MODEL_TYPE}_feature_importance.xlsx"
        ),
        index=False
    )

    print("Feature importance saved.\n")

print("T test predictions saved.")
```

```python
# ==========================================================
# SECTION 3 — TRAIN fO2 MODEL
# composition → fO2
# CV: composition + IW_oof
# Final model: composition + IW_full
# ==========================================================

import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import joblib
import numpy as np
import pandas as pd
import os

print("Preparing physics-informed fO2 model...")

N_SPLITS = 10
USE_IW_FEATURE = True # True=加入IW，False=不加入IW

fO2_train_df = df_fO2_train.copy()
fO2_test_df  = df_fO2_test.copy()

X_base_train = fO2_train_df[BASE_FEATURES].values
y_train = fO2_train_df["fO2"].values

stratify_labels = fO2_train_df[LITHOLOGY_COLUMN].values
test_lithology_labels = fO2_test_df[LITHOLOGY_COLUMN].values

if USE_IW_FEATURE:
    print("Generating IW_oof for CV...")

    skf = StratifiedKFold(
        n_splits=N_SPLITS,
        shuffle=True,
        random_state=RANDOM_STATE
    )

    P_oof = np.zeros(len(fO2_train_df))
    T_oof = np.zeros(len(fO2_train_df))

    for fold, (tr, val) in enumerate(skf.split(X_base_train, stratify_labels)):

        model_P = create_model(MODEL_TYPE, "P")
        model_T = create_model(MODEL_TYPE, "T")

        model_P.fit(X_base_train[tr], fO2_train_df["P"].values[tr])
        model_T.fit(X_base_train[tr], fO2_train_df["T"].values[tr])

        P_oof[val] = model_P.predict(X_base_train[val])
        T_oof[val] = model_T.predict(X_base_train[val])

    fO2_train_df["IW_oof"] = calculate_iw(P_oof, T_oof)

skf = StratifiedKFold(
    n_splits=N_SPLITS,
    shuffle=True,
    random_state=RANDOM_STATE
)

if USE_IW_FEATURE:
    F_FEATURES_CV = BASE_FEATURES + ["IW_oof"]
else:
    F_FEATURES_CV = BASE_FEATURES

X_cv = fO2_train_df[F_FEATURES_CV].values

print("\nStarting 10-fold CV...")

rmse_list = []
mae_list  = []
r2_list   = []

cv_lith_metrics = {
    LITHOLOGY_1_NAME: {"rmse": [], "mae": [], "r2": []},
    LITHOLOGY_2_NAME: {"rmse": [], "mae": [], "r2": []}
}

best_r2 = -np.inf
best_fold_data = None
best_fold_metrics = None

for fold, (tr, val) in enumerate(skf.split(X_cv, stratify_labels)):

    model = create_model(MODEL_TYPE, "fO2")

    model.fit(X_cv[tr], y_train[tr])

    y_val_pred = model.predict(X_cv[val])
    lith_val = stratify_labels[val]

    rmse = np.sqrt(mean_squared_error(y_train[val], y_val_pred))
    mae  = mean_absolute_error(y_train[val], y_val_pred)
    r2   = r2_score(y_train[val], y_val_pred)

    rmse_list.append(rmse)
    mae_list.append(mae)
    r2_list.append(r2)

    print(f"Fold {fold+1}: RMSE={rmse:.4f}, MAE={mae:.4f}, R2={r2:.4f}")

    for lith_label, lith_name in [
        (LITHOLOGY_1_LABEL, LITHOLOGY_1_NAME),
        (LITHOLOGY_2_LABEL, LITHOLOGY_2_NAME)
    ]:
        mask = lith_val == lith_label
        if np.sum(mask) >= 2:
            cv_lith_metrics[lith_name]["rmse"].append(
                np.sqrt(mean_squared_error(y_train[val][mask], y_val_pred[mask]))
            )
            cv_lith_metrics[lith_name]["mae"].append(
                mean_absolute_error(y_train[val][mask], y_val_pred[mask])
            )
            cv_lith_metrics[lith_name]["r2"].append(
                r2_score(y_train[val][mask], y_val_pred[mask])
            )

    if r2 > best_r2:
        best_r2 = r2
        best_fold_data = (y_train[val], y_val_pred, lith_val)
        best_fold_metrics = (rmse, mae, r2)

print("\nCV Results (Mean ± Std):")
print(f"RMSE: {np.mean(rmse_list):.4f} ± {np.std(rmse_list):.4f}")
print(f"MAE : {np.mean(mae_list):.4f} ± {np.std(mae_list):.4f}")
print(f"R2  : {np.mean(r2_list):.4f} ± {np.std(r2_list):.4f}")

print("\nCV Results by Lithology (Mean ± Std):")
for lith_name in [LITHOLOGY_1_NAME, LITHOLOGY_2_NAME]:
    print(f"\n{lith_name}:")
    print(f"  RMSE: {np.mean(cv_lith_metrics[lith_name]['rmse']):.4f} ± {np.std(cv_lith_metrics[lith_name]['rmse']):.4f}")
    print(f"  MAE : {np.mean(cv_lith_metrics[lith_name]['mae']):.4f} ± {np.std(cv_lith_metrics[lith_name]['mae']):.4f}")
    print(f"  R2  : {np.mean(cv_lith_metrics[lith_name]['r2']):.4f} ± {np.std(cv_lith_metrics[lith_name]['r2']):.4f}")

if USE_IW_FEATURE:
    print("\nGenerating IW_full for final training...")

    P_full_train = P_model.predict(X_base_train)
    T_full_train = T_model.predict(X_base_train)

    fO2_train_df["IW_full"] = calculate_iw(P_full_train, T_full_train)
    F_FEATURES_FINAL = BASE_FEATURES + ["IW_full"]
else:
    F_FEATURES_FINAL = BASE_FEATURES

X_train_final = fO2_train_df[F_FEATURES_FINAL].values

print("\nTraining final fO2_model...")

fO2_model = create_model(MODEL_TYPE, "fO2")

fO2_model.fit(X_train_final, y_train)


joblib.dump(
    fO2_model,
    os.path.join(MODEL_PATH, f"fO2_model_{MODEL_TYPE}.pkl")
)

print("fO2 model saved.")

print("\nEvaluating on test set...")

X_base_test = fO2_test_df[BASE_FEATURES].values
y_test = fO2_test_df["fO2"].values

if USE_IW_FEATURE:
    P_test = P_model.predict(X_base_test)
    T_test = T_model.predict(X_base_test)

    fO2_test_df["IW_full"] = calculate_iw(P_test, T_test)

X_test = fO2_test_df[F_FEATURES_FINAL].values

y_test_pred = fO2_model.predict(X_test)

test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
test_mae  = mean_absolute_error(y_test, y_test_pred)
test_r2   = r2_score(y_test, y_test_pred)

print("\nTest Results:")
print(f"RMSE: {test_rmse:.4f}")
print(f"MAE : {test_mae:.4f}")
print(f"R2  : {test_r2:.4f}")

print("\nTest Results by Lithology:")
for lith_label, lith_name in [
    (LITHOLOGY_1_LABEL, LITHOLOGY_1_NAME),
    (LITHOLOGY_2_LABEL, LITHOLOGY_2_NAME)
]:
    mask = test_lithology_labels == lith_label
    y_true_l = y_test[mask]
    y_pred_l = y_test_pred[mask]
    residual_l = y_true_l - y_pred_l
    abs_error_l = np.abs(residual_l)

    print(f"\n{lith_name}:")
    print(f"  N   : {np.sum(mask)}")
    print(f"  RMSE: {np.sqrt(mean_squared_error(y_true_l, y_pred_l)):.4f}")
    print(f"  MAE : {mean_absolute_error(y_true_l, y_pred_l):.4f}")
    print(f"  R2  : {r2_score(y_true_l, y_pred_l):.4f}")
    print(f"  Residual mean ± std : {np.mean(residual_l):.4f} ± {np.std(residual_l):.4f}")
    print(f"  Abs error mean ± std: {np.mean(abs_error_l):.4f} ± {np.std(abs_error_l):.4f}")

import matplotlib as mpl

mpl.rcParams["font.family"] = "Arial"
mpl.rcParams["font.size"] = 11

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

color_1 = "#3D5A80"
color_2 = "#D08C60"
line_color = "gray"

y_true_best, y_pred_best, lith_best = best_fold_data
rmse_best, mae_best, r2_best = best_fold_metrics

mask_1 = lith_best == LITHOLOGY_1_LABEL
mask_2 = lith_best == LITHOLOGY_2_LABEL

axes[0].scatter(y_true_best[mask_1], y_pred_best[mask_1], s=80, color=color_1, alpha=0.7, edgecolors="none", label=LITHOLOGY_1_NAME)
axes[0].scatter(y_true_best[mask_2], y_pred_best[mask_2], s=80, color=color_2, alpha=0.7, edgecolors="none", label=LITHOLOGY_2_NAME)

min_val = min(y_true_best.min(), y_pred_best.min())
max_val = max(y_true_best.max(), y_pred_best.max())

axes[0].plot([min_val, max_val], [min_val, max_val], linestyle="--", color=line_color, linewidth=1.5)

axes[0].set_title("Best CV Fold")
axes[0].set_xlabel("True logfO$_2$")
axes[0].set_ylabel("Predicted logfO$_2$")

axes[0].text(0.05, 0.95,
    f"RMSE = {rmse_best:.3f}\nMAE = {mae_best:.3f}\nR² = {r2_best:.3f}",
    transform=axes[0].transAxes, verticalalignment="top",
    bbox=dict(facecolor="white", edgecolor="none", alpha=0.8))

axes[0].legend(frameon=False, loc="lower right")
axes[0].set_aspect("equal", adjustable="box")
axes[0].grid(False)

mask_1_test = test_lithology_labels == LITHOLOGY_1_LABEL
mask_2_test = test_lithology_labels == LITHOLOGY_2_LABEL

axes[1].scatter(y_test[mask_1_test], y_test_pred[mask_1_test], s=80, color=color_1, alpha=0.7, edgecolors="none", label=LITHOLOGY_1_NAME)
axes[1].scatter(y_test[mask_2_test], y_test_pred[mask_2_test], s=80, color=color_2, alpha=0.7, edgecolors="none", label=LITHOLOGY_2_NAME)

min_val = min(y_test.min(), y_test_pred.min())
max_val = max(y_test.max(), y_test_pred.max())

axes[1].plot([min_val, max_val], [min_val, max_val], linestyle="--", color=line_color, linewidth=1.5)

axes[1].set_title("Test Set")
axes[1].set_xlabel("True logfO$_2$")
axes[1].set_ylabel("Predicted logfO$_2$")

axes[1].text(0.05, 0.95,
    f"RMSE = {test_rmse:.3f}\nMAE = {test_mae:.3f}\nR² = {test_r2:.3f}",
    transform=axes[1].transAxes, verticalalignment="top",
    bbox=dict(facecolor="white", edgecolor="none", alpha=0.8))

axes[1].legend(frameon=False, loc="lower right")
axes[1].set_aspect("equal", adjustable="box")
axes[1].grid(False)

for ax in axes:
    for spine in ax.spines.values():
        spine.set_linewidth(2)

plt.tight_layout()


save_path_png = os.path.join(IMAGE_PATH, "Fig_c_fO2_model.png")
save_path_pdf = os.path.join(IMAGE_PATH, "Fig_c_fO2_model.pdf")

plt.savefig(save_path_png, dpi=600)
plt.savefig(save_path_pdf)

plt.show()

test_result = pd.DataFrame({
    "logfO2_true": y_test,
    "logfO2_pred": y_test_pred,
    "lithology": test_lithology_labels
})

test_result["residual"] = (
    test_result["logfO2_true"] - test_result["logfO2_pred"]
)

test_result["abs_error"] = np.abs(test_result["residual"])

test_result.to_excel(
    os.path.join(RESULT_DIR, "logfO2_test_predictions.xlsx"),
    index=False
)

if MODEL_TYPE in ["xgboost", "rf"]:

    print("\nComputing feature importance...")

    importance = fO2_model.feature_importances_

    feature_names = BASE_FEATURES + ["IW_full"] if USE_IW_FEATURE else BASE_FEATURES

    importance_df = pd.DataFrame({
        "feature": feature_names,
        "importance": importance
    }).sort_values("importance", ascending=False)

    print("\nFeature Importance:")
    print(importance_df)

    plt.figure(figsize=(6, 4))

    plt.barh(importance_df["feature"], importance_df["importance"])
    plt.gca().invert_yaxis()

    plt.xlabel("Importance")
    plt.title(f"{MODEL_TYPE.upper()} Feature Importance")

    plt.tight_layout()


    plt.savefig(
        os.path.join(
            IMAGE_PATH,
            f"{MODEL_TYPE}_feature_importance.png"
        ),
        dpi=300
    )

    plt.show()

    importance_df.to_excel(
        os.path.join(
            RESULT_DIR,
            f"{MODEL_TYPE}_feature_importance.xlsx"
        ),
        index=False
    )

    print("Feature importance saved.\n")

print("SECTION 3 completed.\n")
```

```python
# ==========================================================
# CELL — dIW COMPARISON (TRUE vs PREDICTED) BY LITHOLOGY
# dIW = fO2 - IW
# True dIW:      use true P-T-fO2
# Predicted dIW: use predicted P-T-fO2
# Lithology:
#   1 = Peridotite
#   2 = Eclogite
# ==========================================================

import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import pandas as pd
import os

print("Calculating true and predicted dIW on test set...")

# -------------------------------------------------
# 1️⃣ True IW and True dIW
# -------------------------------------------------
P_true = fO2_test_df["P"].values
T_true = fO2_test_df["T"].values
fO2_true = fO2_test_df["fO2"].values

IW_true = calculate_iw(P_true, T_true)
dIW_true = fO2_true - IW_true

# -------------------------------------------------
# 2️⃣ Predicted IW and Predicted dIW
# -------------------------------------------------
IW_pred = calculate_iw(P_test, T_test)
dIW_pred = y_test_pred - IW_pred

# -------------------------------------------------
# 3️⃣ Lithology
# -------------------------------------------------
lithology_code = fO2_test_df[LITHOLOGY_COLUMN].values

lithology_name = pd.Series(lithology_code).map({
    1: "Peridotite",
    2: "Eclogite"
}).fillna("Unknown").values

# -------------------------------------------------
# 4️⃣ Overall Metrics
# -------------------------------------------------
diw_rmse = np.sqrt(mean_squared_error(dIW_true, dIW_pred))
diw_mae  = mean_absolute_error(dIW_true, dIW_pred)
diw_r2   = r2_score(dIW_true, dIW_pred)

print("\nOverall dIW Test Results:")
print(f"RMSE: {diw_rmse:.4f}")
print(f"MAE : {diw_mae:.4f}")
print(f"R2  : {diw_r2:.4f}")

# -------------------------------------------------
# 5️⃣ Metrics by lithology
# -------------------------------------------------
print("\nMetrics by Lithology:")

for code, name in [(1, "Peridotite"), (2, "Eclogite")]:
    mask = lithology_code == code
    if np.sum(mask) > 1:
        rmse_i = np.sqrt(mean_squared_error(dIW_true[mask], dIW_pred[mask]))
        mae_i  = mean_absolute_error(dIW_true[mask], dIW_pred[mask])
        r2_i   = r2_score(dIW_true[mask], dIW_pred[mask])
        print(f"{name}: n={np.sum(mask)}, RMSE={rmse_i:.4f}, MAE={mae_i:.4f}, R2={r2_i:.4f}")
    elif np.sum(mask) == 1:
        print(f"{name}: n=1, only one sample, R2 unavailable")
    else:
        print(f"{name}: n=0")

# -------------------------------------------------
# 6️⃣ Plot
# -------------------------------------------------
mpl.rcParams["font.family"] = "Arial"
mpl.rcParams["font.size"] = 11

fig, ax = plt.subplots(figsize=(5.8, 5.5))

color_map = {
    1: "#3D5A80",  # Peridotite
    2: "#D08C60"   # Eclogite
}


label_map = {
    1: "Peridotite",
    2: "Eclogite"
}

for code in [1, 2]:
    mask = lithology_code == code
    if np.sum(mask) > 0:
        ax.scatter(
            dIW_true[mask],
            dIW_pred[mask],
            s=80,
            color=color_map[code],
            alpha=0.7,
            edgecolors="none",
            label=label_map[code]
        )

# global axis range
min_val = min(dIW_true.min(), dIW_pred.min())
max_val = max(dIW_true.max(), dIW_pred.max())

pad = 0.5
min_plot = min_val - pad
max_plot = max_val + pad

xline = np.array([min_plot, max_plot])

# 1:1 line
ax.plot(
    xline, xline,
    linestyle="-",
    color="black",
    linewidth=1.5
)

# +2 line
ax.plot(
    xline, xline + 1,
    linestyle="--",
    color="gray",
    linewidth=1.2
)

# -2 line
ax.plot(
    xline, xline - 1,
    linestyle="--",
    color="gray",
    linewidth=1.2
)

ax.set_xlim(min_plot, max_plot)
ax.set_ylim(min_plot, max_plot)

ax.set_title("Test Set (dIW)")
ax.set_xlabel("True dIW")
ax.set_ylabel("Predicted dIW")

ax.text(
    0.05, 0.95,
    f"RMSE = {diw_rmse:.3f}\nMAE = {diw_mae:.3f}\nR² = {diw_r2:.3f}",
    transform=ax.transAxes,
    verticalalignment="top",
    bbox=dict(facecolor="white", edgecolor="none", alpha=0.85)
)

ax.legend(frameon=False, loc="lower right")
ax.set_aspect("equal", adjustable="box")
ax.grid(False)

for spine in ax.spines.values():
    spine.set_linewidth(2)

plt.tight_layout()

# -------------------------------------------------
# 7️⃣ Save figure
# -------------------------------------------------
save_path_png = os.path.join(IMAGE_PATH, "Fig_d_dIW_model_by_lithology.png")
save_path_pdf = os.path.join(IMAGE_PATH, "Fig_d_dIW_model_by_lithology.pdf")

plt.savefig(save_path_png, dpi=600)
plt.savefig(save_path_pdf)
plt.show()

# -------------------------------------------------
# 8️⃣ Save dIW predictions
# -------------------------------------------------
diw_result = pd.DataFrame({
    #"original_index": fO2_test_df["original_index"].values,
    "lithology_code": lithology_code,
    "lithology_name": lithology_name,
    "P_true": P_true,
    "T_true": T_true,
    "fO2_true": fO2_true,
    "IW_true": IW_true,
    "dIW_true": dIW_true,
    "P_pred": P_test,
    "T_pred": T_test,
    "fO2_pred": y_test_pred,
    "IW_pred": IW_pred,
    "dIW_pred": dIW_pred
})

diw_result["residual_dIW"] = diw_result["dIW_true"] - diw_result["dIW_pred"]
diw_result["abs_error_dIW"] = np.abs(diw_result["residual_dIW"])

diw_result.to_excel(
    os.path.join(RESULT_DIR, "dIW_test_predictions_by_lithology.xlsx"),
    index=False
)

print("dIW predictions by lithology saved.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

plt.rcParams.update({
    "font.family": "sans-serif",
    "font.sans-serif": ["Arial", "DejaVu Sans"],
    "font.size": 18,
    "axes.labelsize": 20,
    "axes.titlesize": 21,
    "xtick.labelsize": 16,
    "ytick.labelsize": 16,
    "axes.linewidth": 1.4,
    "xtick.major.width": 1.3,
    "ytick.major.width": 1.3,
    "xtick.major.size": 6,
    "ytick.major.size": 6,
    "pdf.fonttype": 42,
    "ps.fonttype": 42
})

# ------------------------------
# 1. FMQ calculation
# ------------------------------
def calculate_fmq(P_gpa, T_c):
    P_kbar = np.asarray(P_gpa) * 10.0
    T_k = np.asarray(T_c) + 273.15

    fmq = (
        -587474
        + 1584.427 * T_k
        - 203.3164 * T_k * np.log(T_k)
        + 0.092710 * (T_k ** 2)
        + 1810 * P_kbar
    ) / (8.3144 * T_k * np.log(10))

    return fmq

# ------------------------------
# 2. Compute residual
# ------------------------------
plot_df = df_predict_result.copy()

plot_df["FMQ_pred"] = calculate_fmq(
    plot_df["P_pred"],
    plot_df["T_pred"]
)

plot_df["dFMQ2"] = plot_df["logfO2_pred"] - plot_df["FMQ_pred"]
plot_df["residual_dFMQ"] = plot_df["dFMQ2"] - plot_df["dFMQ"]

# 标准化字段
plot_df["type"] = plot_df["type"].astype(str).str.strip().str.lower()
plot_df["OOD_level"] = plot_df["OOD_level"].astype(str).str.strip()

# ------------------------------
# 3. Filter for Peridotite only
# ------------------------------
peridotite_df = plot_df[plot_df["type"] == "peridotite"].copy()
peridotite_df = peridotite_df.reset_index(drop=True)
peridotite_df["sample_order"] = np.arange(1, len(peridotite_df) + 1)

# Non-OOD statistics only
non_ood_df = peridotite_df[peridotite_df["OOD_level"] != "OOD"].copy()
ood_df = peridotite_df[peridotite_df["OOD_level"] == "OOD"].copy()

print("Total samples:", len(plot_df))
print("Total Peridotite samples:", len(peridotite_df))
print("Peridotite + Non-OOD samples:", len(non_ood_df))
print("Peridotite + OOD samples:", len(ood_df))

print("\nResidual summary (Peridotite + Non-OOD only):")
print(non_ood_df["residual_dFMQ"].describe())

# ------------------------------
# 3.1 Agreement / bias metrics
# ------------------------------
residual = non_ood_df["residual_dFMQ"].dropna().values

mean_bias = np.mean(residual)
mae = np.mean(np.abs(residual))
rmse = np.sqrt(np.mean(residual ** 2))

print(f"\nMean bias (Peridotite + Non-OOD): {mean_bias:.4f}")
print(f"Mean Absolute Error, MAE (Peridotite + Non-OOD): {mae:.4f}")
print(f"Root Mean Squared Error, RMSE (Peridotite + Non-OOD): {rmse:.4f}")

# Percentage within ±1.2
within_err_pct = np.mean(np.abs(residual) <= 1.2) * 100
print(f"Percentage within ±1.2 (Peridotite + Non-OOD): {within_err_pct:.2f}%")

# Symmetric interval ±x containing 95% of samples
x_95 = np.percentile(np.abs(residual), 95)
print(f"±{x_95:.4f} contains 95% of samples (Peridotite + Non-OOD)")

# ------------------------------
# 4. Plot: Peridotite only, OOD distinguished
# ------------------------------
fig, ax = plt.subplots(figsize=(11, 6.2))

color_non_ood = "#4C72B0"  # blue
color_ood = "#D62728"      # red

# Plot Non-OOD
if len(non_ood_df) > 0:
    ax.scatter(
        non_ood_df["sample_order"],
        non_ood_df["residual_dFMQ"],
        marker="s",
        s=65,
        facecolors=color_non_ood,
        edgecolors="black",
        linewidths=0.4,
        alpha=0.9,
        label="Non-OOD"
    )

# Plot OOD
if len(ood_df) > 0:
    ax.scatter(
        ood_df["sample_order"],
        ood_df["residual_dFMQ"],
        marker="s",
        s=65,
        facecolors=color_ood,
        edgecolors="black",
        linewidths=0.4,
        alpha=0.9,
        label="OOD"
    )

# Reference lines
ax.axhline(0, linestyle="-", linewidth=1.5, color="black")
ax.axhline(1.2, linestyle="--", linewidth=1.4, color="black")
ax.axhline(-1.2, linestyle="--", linewidth=1.4, color="black")

# Shaded band for ±1.2
x_min, x_max = 1, len(peridotite_df)
ax.fill_between([x_min - 1, x_max + 1], -1.2, 1.2, alpha=0.08)

# Label positions for ±1.2
x_text = max(1, x_max * 0.08)
offset = 0.25
ax.text(x_text, 1.2 + offset, "+1.2", ha="left", va="bottom", fontsize=17)
ax.text(x_text, -1.2 - offset, "-1.2", ha="left", va="top", fontsize=17)

# Labels
ax.set_xlabel("Sample order", labelpad=10)
ax.set_ylabel("Residual (model - reference)", labelpad=10)

# Axis limits
ax.set_ylim(-5, 5)
ax.set_xlim(x_min - 1, x_max + 1)
ax.margins(x=0)

# Full box
for spine in ax.spines.values():
    spine.set_visible(True)
    spine.set_linewidth(1.4)

ax.tick_params(direction="out")

# X ticks
n_samples = len(peridotite_df)
if n_samples > 40:
    step = max(1, n_samples // 10)
    ax.set_xticks(np.arange(1, n_samples + 1, step))
else:
    ax.set_xticks(np.arange(1, n_samples + 1))

# Legend
ax.legend(frameon=False)

plt.tight_layout()

# ------------------------------
# 5. Save
# ------------------------------
os.makedirs(IMAGE_PATH, exist_ok=True)

plt.savefig(
    os.path.join(
        IMAGE_PATH,
        f"{MODEL_TYPE}_dFMQ_residual_peridotite_with_OOD.png"
    ),
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from matplotlib.lines import Line2D

plt.rcParams.update({
    "font.family": "sans-serif",
    "font.sans-serif": ["Arial", "DejaVu Sans"],
    "font.size": 18,
    "axes.labelsize": 20,
    "axes.titlesize": 21,
    "xtick.labelsize": 16,
    "ytick.labelsize": 16,
    "axes.linewidth": 1.4,
    "xtick.major.width": 1.3,
    "ytick.major.width": 1.3,
    "xtick.major.size": 6,
    "ytick.major.size": 6,
    "pdf.fonttype": 42,
    "ps.fonttype": 42
})

# ------------------------------
# 1. FMQ calculation
# ------------------------------
def calculate_fmq(P_gpa, T_c):
    P_kbar = np.asarray(P_gpa) * 10.0
    T_k = np.asarray(T_c) + 273.15

    fmq = (
        -587474
        + 1584.427 * T_k
        - 203.3164 * T_k * np.log(T_k)
        + 0.092710 * (T_k ** 2)
        + 1810 * P_kbar
    ) / (8.3144 * T_k * np.log(10))

    return fmq

# ------------------------------
# 2. Compute residual
# ------------------------------
plot_df = df_predict_result.copy()

plot_df["FMQ_pred"] = calculate_fmq(
    plot_df["P_pred"],
    plot_df["T_pred"]
)

plot_df["dFMQ2"] = plot_df["logfO2_pred"] - plot_df["FMQ_pred"]
plot_df["residual_dFMQ"] = plot_df["dFMQ2"] - plot_df["dFMQ"]

# 标准化 type
plot_df["type"] = plot_df["type"].astype(str).str.strip().str.lower()

# Marker mapping
marker_map = {
    "peridotite": "s",
    "eclogite": "o"
}
default_marker = "^"

# Color mapping
color_in = "#4C72B0"
color_ood = "#D62728"

# ------------------------------
# 3. Summary
# ------------------------------
in_df = plot_df[plot_df["OOD_level"] == "In-Distribution"]
ood_df = plot_df[plot_df["OOD_level"] == "OOD"]

print("Total samples:", len(plot_df))
print("In-Distribution samples:", len(in_df))
print("OOD samples:", len(ood_df))

print("\nResidual summary (In-Distribution only):")
print(in_df["residual_dFMQ"].describe())

within_err = in_df["residual_dFMQ"].between(-1.2, 1.2).mean() * 100
print(f"\nWithin ±1.2 (In-Distribution): {within_err:.2f}%")

# ------------------------------
# 4. Plot（关键修改：x轴）
# ------------------------------
fig, ax = plt.subplots(figsize=(11, 6.2))

x_col = "Fe3+/Fe"   # ⭐ 核心修改

for sample_type in plot_df["type"].dropna().unique():
    marker = marker_map.get(sample_type, default_marker)

    mask_type = plot_df["type"] == sample_type
    mask_in = mask_type & (plot_df["OOD_level"] == "In-Distribution")
    mask_ood = mask_type & (plot_df["OOD_level"] == "OOD")

    if mask_in.any():
        ax.scatter(
            plot_df.loc[mask_in, x_col],
            plot_df.loc[mask_in, "residual_dFMQ"],
            marker=marker,
            s=65,
            facecolors=color_in,
            edgecolors="black",
            linewidths=0.4,
            alpha=0.9
        )

    if mask_ood.any():
        ax.scatter(
            plot_df.loc[mask_ood, x_col],
            plot_df.loc[mask_ood, "residual_dFMQ"],
            marker=marker,
            s=65,
            facecolors=color_ood,
            edgecolors="black",
            linewidths=0.4,
            alpha=0.9
        )

# Reference lines
ax.axhline(0, linestyle="-", linewidth=1.5)
ax.axhline(1.2, linestyle="--", linewidth=1.4)
ax.axhline(-1.2, linestyle="--", linewidth=1.4)

# 阴影（用当前x范围）
xmin = plot_df[x_col].min()
xmax = plot_df[x_col].max()

ax.fill_between([xmin-0.02, xmax+0.02], -1.2, 1.2, alpha=0.08)

# 标签位置（左侧）
x_text = xmin + (xmax - xmin) * 0.95
offset = 0.25

ax.text(x_text, 1.2 + offset, "+1.2", ha="left", va="bottom", fontsize=17)
ax.text(x_text, -1.2 - offset, "-1.2", ha="left", va="top", fontsize=17)

# Labels
ax.set_xlabel("Fe3+/Fe", labelpad=10)
ax.set_ylabel("Residual", labelpad=10)

# 轴范围
ax.set_ylim(-5, 5)
ax.set_xlim(xmin-0.02, xmax+0.02)

# 边框
for spine in ax.spines.values():
    spine.set_visible(True)
    spine.set_linewidth(1.4)

ax.tick_params(direction="out")

# Legend（只保留岩性）
legend_elements = [
    Line2D([0], [0], marker='s', color='black',
           markerfacecolor='white', markeredgecolor='black',
           markersize=9, linestyle='None', label='Peridotite'),

    Line2D([0], [0], marker='o', color='black',
           markerfacecolor='white', markeredgecolor='black',
           markersize=9, linestyle='None', label='Eclogite')
]

ax.legend(handles=legend_elements, frameon=False)

plt.tight_layout()

# ------------------------------
# 5. Save
# ------------------------------
os.makedirs(IMAGE_PATH, exist_ok=True)

plt.savefig(
    os.path.join(
        IMAGE_PATH,
        f"{MODEL_TYPE}_dFMQ_vs_Fe3Fe.png"
    ),
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
# ==============================================================
# FULL PER-SAMPLE SHAP MODULE
# Stable version （这个是最终完整版）
# ==============================================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
import shapiq

warnings.filterwarnings("ignore")

print("\n========== STARTING FULL PER-SAMPLE SHAP ==========\n")

SHAP_TARGET_MODEL = "ALL"# "P" / "T" / "fO2" / "ALL"
SHAP_BUDGET =500

SHAP_ROOT = os.path.join(RESULT_DIR, "SHAP_results")
os.makedirs(SHAP_ROOT, exist_ok=True)


# ==============================================================
# DATA PREPARATION
# ==============================================================

def prepare_data(model_name):

    if model_name == "P":
        train_df = pd.concat([df_P_T_train, df_fO2_train], ignore_index=True)
        test_df = pd.concat([df_P_T_test, df_fO2_test], ignore_index=True)
        feature_list = BASE_FEATURES
        target = "P"

    elif model_name == "T":
        train_df = pd.concat([df_P_T_train, df_fO2_train], ignore_index=True)
        test_df = pd.concat([df_P_T_test, df_fO2_test], ignore_index=True)
        feature_list = BASE_FEATURES
        target = "T"

    elif model_name == "fO2":
        train_df = df_fO2_train.copy()
        test_df = df_fO2_test.copy()
        feature_list = BASE_FEATURES + ["IW_full"]
        target = "fO2"

    else:
        raise ValueError("Unknown model_name.")

    return train_df, test_df, feature_list, target


# ==============================================================
# TRAIN LOCAL MODELS FOR fO2
# ==============================================================

def train_local_models_for_sample():

    P_train_df = pd.concat([df_P_T_train, df_fO2_train], ignore_index=True)
    X_P = P_train_df[BASE_FEATURES].values
    y_P = P_train_df["P"].values

    P_model_local = create_model(MODEL_TYPE, "P")
    P_model_local.fit(X_P, y_P)

    T_train_df = pd.concat([df_P_T_train, df_fO2_train], ignore_index=True)
    X_T = T_train_df[BASE_FEATURES].values
    y_T = T_train_df["T"].values

    T_model_local = create_model(MODEL_TYPE, "T")
    T_model_local.fit(X_T, y_T)

    fO2_train_df_local = df_fO2_train.copy()
    X_f_base = fO2_train_df_local[BASE_FEATURES].values
    y_f = fO2_train_df_local["fO2"].values

    P_full_train = P_model_local.predict(X_f_base)
    T_full_train = T_model_local.predict(X_f_base)

    fO2_train_df_local["IW_full"] = calculate_iw(P_full_train, T_full_train)

    fO2_feature_list = BASE_FEATURES + ["IW_full"]
    X_f_train = fO2_train_df_local[fO2_feature_list].values

    fO2_model_local = create_model(MODEL_TYPE, "fO2")
    fO2_model_local.fit(X_f_train, y_f)

    return P_model_local, T_model_local, fO2_model_local, fO2_train_df_local


# ==============================================================
# EXTRACT SHAP VALUES
# ==============================================================

def extract_sv_from_dict_values(explanation, n_features):

    values_dict = explanation.dict_values
    baseline = float(values_dict[()])

    shap_vals = []
    for j in range(n_features):
        shap_vals.append(float(values_dict[(j,)]))

    shap_vals = np.array(shap_vals)
    return baseline, shap_vals, values_dict


# ==============================================================
# MAIN SHAP FUNCTION
# ==============================================================

def run_shap(model_name):

    print(f"\nRunning SHAP for {model_name}")

    train_df, test_df, feature_list, target = prepare_data(model_name)

    shap_values_all = []
    baseline_all = []
    prediction_all = []
    reconstructed_all = []
    difference_all = []

    feature_values_all = []   # ★ 新增：缓存 feature value

    extra_info = []
    raw_dict_values_all = []

    for i in range(len(test_df)):

        if i % 10 == 0:
            print(f"Explaining sample {i+1}/{len(test_df)}")

        # ------------------------------------------------------
        # Train local models
        # ------------------------------------------------------

        if model_name in ["P", "T"]:

            train_local = pd.concat([df_P_T_train, df_fO2_train], ignore_index=True)

            X_train = train_local[BASE_FEATURES].values
            y_train = train_local[target].values

            model_local = create_model(MODEL_TYPE, target)
            model_local.fit(X_train, y_train)

            x_single = test_df.iloc[i:i+1][BASE_FEATURES].values

        elif model_name == "fO2":

            P_model_local, T_model_local, fO2_model_local, fO2_train_local = \
                train_local_models_for_sample()

            X_train = fO2_train_local[BASE_FEATURES + ["IW_full"]].values
            y_train = fO2_train_local["fO2"].values

            x_base_single = test_df.iloc[i:i+1][BASE_FEATURES].values

            P_single = float(P_model_local.predict(x_base_single)[0])
            T_single = float(T_model_local.predict(x_base_single)[0])
            IW_single = float(calculate_iw(np.array([P_single]), np.array([T_single]))[0])

            x_single_df = test_df.iloc[i:i+1][BASE_FEATURES].copy()
            x_single_df["IW_full"] = IW_single

            x_single = x_single_df[BASE_FEATURES + ["IW_full"]].values

            model_local = fO2_model_local

            extra_info.append({
                "P_pred_local": P_single,
                "T_pred_local": T_single,
                "IW_pred_local": IW_single
            })

        # ------------------------------------------------------
        # 保存 feature value
        # ------------------------------------------------------

        feature_values_all.append(x_single.flatten())

        # ------------------------------------------------------
        # Prediction
        # ------------------------------------------------------

        pred = float(model_local.predict(x_single)[0])

        average_prediction = float(np.mean(model_local.predict(X_train)))

        explainer = shapiq.Explainer(
            model=model_local,
            data=X_train,
            labels=y_train,
            index="SV",
            max_order=1,
            empty_prediction=average_prediction,
        )

        explanation = explainer.explain(
            x_single,
            budget=SHAP_BUDGET
        )

        baseline, shap_vals, values_dict = extract_sv_from_dict_values(
            explanation,
            n_features=len(feature_list)
        )

        shap_sum = float(shap_vals.sum())
        reconstructed = baseline + shap_sum
        difference = pred - reconstructed

        shap_values_all.append(shap_vals)
        baseline_all.append(baseline)
        prediction_all.append(pred)
        reconstructed_all.append(reconstructed)
        difference_all.append(difference)
        raw_dict_values_all.append(str(values_dict))

        del explainer
        del model_local

    # ----------------------------------------------------------
    # Convert arrays
    # ----------------------------------------------------------

    shap_values_all = np.vstack(shap_values_all)
    feature_values_all = np.vstack(feature_values_all)   # ★ 新增

    baseline_all = np.array(baseline_all)
    prediction_all = np.array(prediction_all)
    reconstructed_all = np.array(reconstructed_all)
    difference_all = np.array(difference_all)

    # ----------------------------------------------------------
    # Save results
    # ----------------------------------------------------------

    shap_dir = os.path.join(SHAP_ROOT, model_name)
    os.makedirs(shap_dir, exist_ok=True)

    full_df = pd.DataFrame(
        shap_values_all,
        columns=feature_list
    )

    # ★ 可选增强：保存 feature value
    feature_df = pd.DataFrame(
        feature_values_all,
        columns=[f"{f}_value" for f in feature_list]
    )

    full_df = pd.concat([feature_df, full_df], axis=1)

    full_df["baseline"] = baseline_all
    full_df["prediction"] = prediction_all
    full_df["difference"] = difference_all

    full_df.to_excel(
        os.path.join(shap_dir, "shap_full_results.xlsx"),
        index=False
    )

    # ----------------------------------------------------------
    # Dependence plots
    # ----------------------------------------------------------

    if LITHOLOGY_COLUMN in test_df.columns:

        for label, name in [
            (LITHOLOGY_1_LABEL, LITHOLOGY_1_NAME),
            (LITHOLOGY_2_LABEL, LITHOLOGY_2_NAME)
        ]:

            mask = test_df[LITHOLOGY_COLUMN] == label
            if mask.sum() == 0:
                continue

            shap_lith = shap_values_all[mask]

            X_lith = feature_values_all[mask]   # ★ 修复：使用真实 feature

            dep_df = pd.DataFrame()

            for j, feature in enumerate(feature_list):

                plt.figure(figsize=(6,4))

                plt.scatter(
                    X_lith[:, j],
                    shap_lith[:, j],
                    alpha=0.6
                )

                plt.xlabel(feature)
                plt.ylabel("SHAP value")
                plt.title(f"{model_name} Dependence — {feature} ({name})")

                plt.tight_layout()

                lith_dir = os.path.join(shap_dir, name)
                os.makedirs(lith_dir, exist_ok=True)

                plt.savefig(
                    os.path.join(lith_dir, f"dependence_{feature}.png"),
                    dpi=300
                )

                plt.close()

                dep_df[f"{feature}_value"] = X_lith[:, j]
                dep_df[f"{feature}_shap"] = shap_lith[:, j]

            dep_df.to_excel(
                os.path.join(lith_dir, "dependence_raw.xlsx"),
                index=False
            )


# ==============================================================
# EXECUTION
# ==============================================================

if SHAP_TARGET_MODEL == "ALL":
    for m in ["P","T","fO2"]:
        run_shap(m)
else:
    run_shap(SHAP_TARGET_MODEL)

print("\n========== FULL PER-SAMPLE SHAP COMPLETE ==========\n")
```

```python
# ==========================================================
# FULL OXYBAROMETER BENCHMARK (TABLE FORMAT OUTPUT)
# Pipeline: composition → P → T → IW → fO2
# Algorithms: TabPFN / XGBoost / RF
# Metrics: 10-fold CV(mean ± std) + Test
# Output table format:
# Model | Target | CV_R2±std | CV_RMSE±std | Test_R2 | Test_RMSE
# ==========================================================

import numpy as np
import pandas as pd
import os
import time

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

print("\n==============================================")
print("Running FULL oxybarometer benchmark")
print("==============================================\n")

N_SPLITS = 10

algorithms = ["tabpfn","xgboost","rf"]

table_rows = []

# ==========================================================
# DATASETS
# ==========================================================

PT_train = pd.concat(
    [df_P_T_train, df_fO2_train],
    ignore_index=True
)

PT_test = pd.concat(
    [df_P_T_test, df_fO2_test],
    ignore_index=True
)

fO2_train = df_fO2_train.copy()
fO2_test  = df_fO2_test.copy()

# ==========================================================
# MAIN LOOP
# ==========================================================

for algo in algorithms:

    print("\n==============================")
    print("Algorithm:",algo)
    print("==============================")

    # ------------------------------------------------------
    # P MODEL CV
    # ------------------------------------------------------

    X_PT = PT_train[BASE_FEATURES]
    strat_labels_PT = PT_train[LITHOLOGY_COLUMN]

    y_P = PT_train["P"]

    skf_PT = StratifiedKFold(
        n_splits=N_SPLITS,
        shuffle=True,
        random_state=RANDOM_STATE
    )

    rmse_list = []
    r2_list = []

    for tr,val in skf_PT.split(X_PT,strat_labels_PT):

        model = create_model(algo,"P")

        if algo == "tabpfn":
            model.fit(X_PT.iloc[tr].values,y_P.iloc[tr].values)
            pred = model.predict(X_PT.iloc[val].values)
        else:
            model.fit(X_PT.iloc[tr],y_P.iloc[tr])
            pred = model.predict(X_PT.iloc[val])

        rmse_list.append(
            np.sqrt(mean_squared_error(y_P.iloc[val],pred))
        )

        r2_list.append(
            r2_score(y_P.iloc[val],pred)
        )

    cv_rmse_mean = np.mean(rmse_list)
    cv_rmse_std  = np.std(rmse_list)

    cv_r2_mean = np.mean(r2_list)
    cv_r2_std  = np.std(r2_list)

    # ---- Final P model ----

    P_model = create_model(algo,"P")

    if algo == "tabpfn":
        P_model.fit(X_PT.values,y_P.values)
        P_pred = P_model.predict(PT_test[BASE_FEATURES].values)
    else:
        P_model.fit(X_PT,y_P)
        P_pred = P_model.predict(PT_test[BASE_FEATURES])

    P_test_rmse = np.sqrt(mean_squared_error(PT_test["P"],P_pred))
    P_test_r2   = r2_score(PT_test["P"],P_pred)

    table_rows.append({

        "Model":algo,
        "Target":"P",

        "CV_R2":f"{cv_r2_mean:.2f}±{cv_r2_std:.2f}",
        "CV_RMSE":f"{cv_rmse_mean:.2f}±{cv_rmse_std:.2f}",

        "Test_R2":round(P_test_r2,2),
        "Test_RMSE":round(P_test_rmse,2)
    })

    # ------------------------------------------------------
    # T MODEL CV
    # ------------------------------------------------------

    y_T = PT_train["T"]

    rmse_list = []
    r2_list = []

    for tr,val in skf_PT.split(X_PT,strat_labels_PT):

        model = create_model(algo,"T")

        if algo == "tabpfn":
            model.fit(X_PT.iloc[tr].values,y_T.iloc[tr].values)
            pred = model.predict(X_PT.iloc[val].values)
        else:
            model.fit(X_PT.iloc[tr],y_T.iloc[tr])
            pred = model.predict(X_PT.iloc[val])

        rmse_list.append(
            np.sqrt(mean_squared_error(y_T.iloc[val],pred))
        )

        r2_list.append(
            r2_score(y_T.iloc[val],pred)
        )

    cv_rmse_mean = np.mean(rmse_list)
    cv_rmse_std  = np.std(rmse_list)

    cv_r2_mean = np.mean(r2_list)
    cv_r2_std  = np.std(r2_list)

    # ---- Final T model ----

    T_model = create_model(algo,"T")

    if algo == "tabpfn":
        T_model.fit(X_PT.values,y_T.values)
        T_pred = T_model.predict(PT_test[BASE_FEATURES].values)
    else:
        T_model.fit(X_PT,y_T)
        T_pred = T_model.predict(PT_test[BASE_FEATURES])

    T_test_rmse = np.sqrt(mean_squared_error(PT_test["T"],T_pred))
    T_test_r2   = r2_score(PT_test["T"],T_pred)

    table_rows.append({

        "Model":algo,
        "Target":"T",

        "CV_R2":f"{cv_r2_mean:.2f}±{cv_r2_std:.2f}",
        "CV_RMSE":f"{cv_rmse_mean:.2f}±{cv_rmse_std:.2f}",

        "Test_R2":round(T_test_r2,2),
        "Test_RMSE":round(T_test_rmse,2)
    })

    # ======================================================
    # fO2 MODEL CV (WITH IW_oof)
    # ======================================================

    print("Generating IW_oof...")

    X_base = fO2_train[BASE_FEATURES].values
    strat_labels = fO2_train[LITHOLOGY_COLUMN].values

    skf = StratifiedKFold(
        n_splits=N_SPLITS,
        shuffle=True,
        random_state=RANDOM_STATE
    )

    P_oof = np.zeros(len(fO2_train))
    T_oof = np.zeros(len(fO2_train))

    for tr,val in skf.split(X_base,strat_labels):

        model_P = create_model(algo,"P")
        model_T = create_model(algo,"T")

        model_P.fit(X_base[tr],fO2_train["P"].values[tr])
        model_T.fit(X_base[tr],fO2_train["T"].values[tr])

        P_oof[val] = model_P.predict(X_base[val])
        T_oof[val] = model_T.predict(X_base[val])

    fO2_train["IW_oof"] = calculate_iw(P_oof,T_oof)

    X_cv = fO2_train[BASE_FEATURES + ["IW_oof"]].values
    y_fO2 = fO2_train["fO2"].values

    rmse_list = []
    r2_list = []

    for tr,val in skf.split(X_cv,strat_labels):

        model = create_model(algo,"fO2")

        model.fit(X_cv[tr],y_fO2[tr])

        pred = model.predict(X_cv[val])

        rmse_list.append(
            np.sqrt(mean_squared_error(y_fO2[val],pred))
        )

        r2_list.append(
            r2_score(y_fO2[val],pred)
        )

    cv_rmse_mean = np.mean(rmse_list)
    cv_rmse_std  = np.std(rmse_list)

    cv_r2_mean = np.mean(r2_list)
    cv_r2_std  = np.std(r2_list)

    # ------------------------------------------------------
    # FINAL fO2 MODEL
    # ------------------------------------------------------

    P_full = P_model.predict(fO2_train[BASE_FEATURES])
    T_full = T_model.predict(fO2_train[BASE_FEATURES])

    fO2_train["IW_full"] = calculate_iw(P_full,T_full)

    X_final = fO2_train[BASE_FEATURES + ["IW_full"]]

    fO2_model = create_model(algo,"fO2")

    fO2_model.fit(X_final,fO2_train["fO2"])

    # ------------------------------------------------------
    # TEST PIPELINE
    # ------------------------------------------------------

    P_test = P_model.predict(fO2_test[BASE_FEATURES])
    T_test = T_model.predict(fO2_test[BASE_FEATURES])

    fO2_test["IW_full"] = calculate_iw(P_test,T_test)

    X_test = fO2_test[BASE_FEATURES + ["IW_full"]]

    y_pred = fO2_model.predict(X_test)

    test_rmse = np.sqrt(mean_squared_error(fO2_test["fO2"],y_pred))
    test_r2   = r2_score(fO2_test["fO2"],y_pred)

    table_rows.append({

        "Model":algo,
        "Target":"fO2",

        "CV_R2":f"{cv_r2_mean:.2f}±{cv_r2_std:.2f}",
        "CV_RMSE":f"{cv_rmse_mean:.2f}±{cv_rmse_std:.2f}",

        "Test_R2":round(test_r2,2),
        "Test_RMSE":round(test_rmse,2)
    })

# ==========================================================
# BUILD FINAL TABLE
# ==========================================================

benchmark_table = pd.DataFrame(table_rows)

print("\nFinal Benchmark Table:\n")
print(benchmark_table)

output_path = os.path.join(
    RESULT_DIR,
    "oxybarometer_benchmark_table.xlsx"
)

benchmark_table.to_excel(output_path,index=False)

print("\nSaved table to:",output_path)

print("\n==============================================")
print("Benchmark completed.")
print("==============================================\n")
```

```python
# ==========================================================
# GRID SEARCH MODULE
# ==========================================================

import os
import numpy as np
import pandas as pd

from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor


# ----------------------------------------------------------
# CPU CONTROL (VERY IMPORTANT)
# ----------------------------------------------------------

MAX_CPU = 8   # 推荐 Ryzen 7940HX 使用 6–10

os.environ["OMP_NUM_THREADS"] = "1"
os.environ["MKL_NUM_THREADS"] = "1"
os.environ["OPENBLAS_NUM_THREADS"] = "1"


# ----------------------------------------------------------
# MAIN FUNCTION
# ----------------------------------------------------------

def run_grid_search(
        target,
        model_type,
        train_df,
        BASE_FEATURES,
        LITHOLOGY_COLUMN,
        RANDOM_STATE,
        calculate_iw=None,
        best_P_params=None,
        best_T_params=None,
        n_splits=5):

    """
    target: "P" / "T" / "fO2"
    model_type: "xgboost" / "rf"
    """

    print("\n======================================")
    print(f"Grid Search | Target: {target} | Model: {model_type}")
    print("======================================")

    df = train_df.copy()
    stratify_labels = df[LITHOLOGY_COLUMN]

    # ======================================================
    # Feature preparation
    # ======================================================

    if target == "fO2":

        if best_P_params is None or best_T_params is None:
            raise ValueError(
                "For fO2 tuning you must provide best_P_params and best_T_params"
            )

        print("Generating IW_oof...")

        X_base = df[BASE_FEATURES].values

        P_oof = np.zeros(len(df))
        T_oof = np.zeros(len(df))

        skf_oof = StratifiedKFold(
            n_splits=n_splits,
            shuffle=True,
            random_state=RANDOM_STATE
        )

        for tr, val in skf_oof.split(X_base, stratify_labels):

            if model_type == "xgboost":

                P_model = XGBRegressor(
                    objective="reg:squarederror",
                    random_state=RANDOM_STATE,
                    n_jobs=1,
                    **best_P_params
                )

                T_model = XGBRegressor(
                    objective="reg:squarederror",
                    random_state=RANDOM_STATE,
                    n_jobs=1,
                    **best_T_params
                )

            else:

                P_model = RandomForestRegressor(
                    random_state=RANDOM_STATE,
                    n_jobs=1,
                    **best_P_params
                )

                T_model = RandomForestRegressor(
                    random_state=RANDOM_STATE,
                    n_jobs=1,
                    **best_T_params
                )

            P_model.fit(X_base[tr], df["P"].values[tr])
            T_model.fit(X_base[tr], df["T"].values[tr])

            P_oof[val] = P_model.predict(X_base[val])
            T_oof[val] = T_model.predict(X_base[val])

        df["IW_oof"] = calculate_iw(P_oof, T_oof)

        X = df[BASE_FEATURES + ["IW_oof"]]
        y = df["fO2"]

    else:

        X = df[BASE_FEATURES]
        y = df[target]

    # ======================================================
    # Model selection
    # ======================================================

    if model_type == "xgboost":

        model = XGBRegressor(
            random_state=RANDOM_STATE,
            tree_method="hist",
            objective="reg:squarederror",
            n_jobs=1
        )

        param_grid_P = {
            "n_estimators":[3000],
            "min_child_weight":[2],
            "max_depth":[8],
            "learning_rate":[0.01],
            "subsample":[0.5],
            "colsample_bytree":[0.9],
            "gamma":[0]
        }

        param_grid_T = {
            "n_estimators":[3000],
            "min_child_weight":[3],
            "max_depth":[14],
            "learning_rate":[0.01],
            "subsample":[0.5],
            "colsample_bytree":[1],
            "gamma":[0.2]
        }

        param_grid_fO2 = {
            "n_estimators":[2000],
            "min_child_weight":[4],
            "max_depth":[6],
            "learning_rate":[0.05],
            "subsample":[0.9],
            "colsample_bytree":[0.6],
            "gamma":[0.2]
        }

    elif model_type == "rf":

        model = RandomForestRegressor(
            random_state=RANDOM_STATE,
            n_jobs=1
        )

        param_grid_P = {
            "n_estimators": [400],
            "max_depth": [15],
            "min_samples_split": [2],
            "min_samples_leaf": [1],
            "max_features": [0.6]
        }

        param_grid_T = {
            "n_estimators": [2000],
            "max_depth": [23],
            "min_samples_split": [2],
            "min_samples_leaf": [1],
            "max_features": [0.6]
        }

        param_grid_fO2 = {
            "n_estimators": [1500],
            "max_depth": [13],
            "min_samples_split": [2],
            "min_samples_leaf": [1],
            "max_features": [0.8]
        }

    else:
        raise ValueError("model_type must be 'xgboost' or 'rf'")

    # select correct grid

    if target == "P":
        param_grid = param_grid_P

    elif target == "T":
        param_grid = param_grid_T

    else:
        param_grid = param_grid_fO2

    # ======================================================
    # Print computation size
    # ======================================================

    total_candidates = 1
    for v in param_grid.values():
        total_candidates *= len(v)

    total_fits = total_candidates * n_splits

    print(f"\nParameter combinations: {total_candidates}")
    print(f"CV folds: {n_splits}")
    print(f"Total fits: {total_fits}")
    print(f"CPU used: {MAX_CPU}")

    # ======================================================
    # Grid Search
    # ======================================================

    skf = StratifiedKFold(
        n_splits=n_splits,
        shuffle=True,
        random_state=RANDOM_STATE
    )

    # 使用 lithology 进行 stratified split
    cv_splits = list(skf.split(X, stratify_labels))

    grid = GridSearchCV(

        estimator=model,
        param_grid=param_grid,

        cv=cv_splits,   # 使用自定义 split

        scoring="r2",

        n_jobs=MAX_CPU,
        verbose=2
    )

    grid.fit(X, y)

    print("\nBest Parameters:")
    print(grid.best_params_)

    print("\nBest CV R2:")
    print(grid.best_score_)

    best_model = grid.best_estimator_

    return best_model, grid.best_params_, grid.best_score_
```

```python
P_model, best_P_params, score_P = run_grid_search(

    target="P",
    model_type="rf",#xgboost

    train_df=pd.concat([df_P_T_train, df_fO2_train]),

    BASE_FEATURES=BASE_FEATURES,
    LITHOLOGY_COLUMN=LITHOLOGY_COLUMN,
    RANDOM_STATE=RANDOM_STATE
)
```

```python
fO2_model, best_fO2_params, score_fO2 = run_grid_search(

    target="fO2",
    model_type="rf",

    train_df=df_fO2_train,

    BASE_FEATURES=BASE_FEATURES,
    LITHOLOGY_COLUMN=LITHOLOGY_COLUMN,
    RANDOM_STATE=RANDOM_STATE,

    calculate_iw=calculate_iw,
    best_P_params=best_P_params,
    best_T_params=best_T_params
)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ===== 简化 IW（你提供的）=====
def iw_simple(P, T):
    T_kelvin = T + 273.15
    return (
        14.07
        - 28784 / T_kelvin
        - 2.04 * np.log10(T_kelvin)
        + 530 * P / T_kelvin
        + 0.03 * P
    )

# ===== 完整 IW（高压版本，已向量化）=====
def iw_full(P, T):
    P = np.asarray(P)
    T = np.asarray(T)
    T_kelvin = T + 273.15

    x0, x1, x2 = -18.64, 0.04359, -5.069e-6
    P_transition = x0 + x1 * T_kelvin + x2 * T_kelvin**2

    # fcc/bcc
    a = 6.844864 + 1.175691e-1 * P + 1.143873e-3 * P**2
    b = 5.791364e-4 - 2.891434e-4 * P - 2.737171e-7 * P**2
    c = (-7.971469e-5 + 3.198005e-5 * P
         + 1.059554e-10 * P**3 + 2.014461e-7 * np.sqrt(P))
    d = (-2.769002e4 + 5.285977e2 * P - 2.919275 * P**2)

    iw_fcc = a + b*T_kelvin + c*T_kelvin*np.log(T_kelvin) + d/T_kelvin

    # hcp
    e = 8.463095 - 3.000307e-3 * P + 7.213445e-5 * P**2
    f = 1.148738e-3 - 9.352312e-5 * P + 5.161592e-7 * P**2
    g = (-7.448624e-4 - 6.329325e-6 * P
         - 1.407339e-10 * P**3 + 1.830014e-4 * np.sqrt(P))
    h = (-2.782082e4 + 5.285977e2 * P - 8.473231e-1 * P**2)

    iw_hcp = e + f*T_kelvin + g*T_kelvin*np.log(T_kelvin) + h/T_kelvin

    return np.where(P <= P_transition, iw_fcc, iw_hcp)

# ===== 测试范围 =====
T = np.linspace(800, 2000, 200)   # °C
P = 20.0                           # 固定压力 1 GPa

iw1 = iw_simple(P, T)
iw2 = iw_full(P, T)

# ===== 差值 =====
diff = iw2 - iw1

# ===== 打印简单统计 =====
print("Difference (full - simple):")
print(f"Mean: {np.mean(diff):.3f}")
print(f"Std : {np.std(diff):.3f}")
print(f"Min : {np.min(diff):.3f}")
print(f"Max : {np.max(diff):.3f}")

# ===== 画图 =====
plt.figure(figsize=(8,5))

plt.plot(T, iw1, label="IW simple", linewidth=2)
plt.plot(T, iw2, label="IW full", linewidth=2)

plt.xlabel("Temperature (°C)")
plt.ylabel("log fO2")
plt.legend()
plt.title(f"IW comparison at {P} GPa")
plt.grid()

plt.show()

# ===== 差值图 =====
plt.figure(figsize=(8,5))

plt.plot(T, diff, color="red", linewidth=2)
plt.axhline(0, linestyle="--")

plt.xlabel("Temperature (°C)")
plt.ylabel("ΔIW (full - simple)")
plt.title("Difference between IW models")
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# =========================================================
# 1) Frost (1991) Fig.1: FMQ, 573–1200 °C
#
# 表达式：
# log fO2 = A/T + B + C*(P-1)/T
#
# 参数：
# A = -25096.3
# B = 8.735
# C = 0.110
#
# 单位：
# T in K
# P in bar
# =========================================================
def fmq_frost1991(P, T):
    """
    FMQ from Frost (1991), Fig.1, 573–1200 °C

    Parameters
    ----------
    P : float or array-like
        Pressure in GPa
    T : float or array-like
        Temperature in °C

    Returns
    -------
    logfO2 : float or ndarray
    """
    P = np.asarray(P)
    T = np.asarray(T)

    T_kelvin = T + 273.15
    P_bar = P * 1e4  # 1 GPa = 10,000 bar

    A = -25096.3
    B = 8.735
    C = 0.110

    return A / T_kelvin + B + C * (P_bar - 1.0) / T_kelvin


# =========================================================
# 2) O'Neill (1987a)
#
# log fO2 = 82.75 + 0.00484*T - 30681/T
#           - 24.45*log10(T) + 940*P/T - 0.02*P
#
# 单位：
# T in K
# P in GPa
# =========================================================
def fmq_oneill1987(P, T):
    """
    FMQ from O'Neill (1987a)

    Parameters
    ----------
    P : float or array-like
        Pressure in GPa
    T : float or array-like
        Temperature in °C

    Returns
    -------
    logfO2 : float or ndarray
    """
    P = np.asarray(P)
    T = np.asarray(T)

    T_kelvin = T + 273.15

    return (
        82.75
        + 0.00484 * T_kelvin
        - 30681.0 / T_kelvin
        - 24.45 * np.log10(T_kelvin)
        + 940.0 * P / T_kelvin
        - 0.02 * P
    )


# =========================================================
# 3) Holland & Powell pressure-corrected FMQ
#
# 图中公式：
# Log fO2(FMQ) =
# (-587474 + 1584.427*T - 203.3164*T*ln(T) + 0.09271*T^2 + 1810*P)
# / (8.3144*T*ln(10))
#
# 单位：
# T in K
# P in kbar
#
# 输入这里统一用：
# T in °C
# P in GPa
# =========================================================
def fmq_hp2011(P, T):
    """
    FMQ using the Holland & Powell-style expression shown in the figure

    Parameters
    ----------
    P : float or array-like
        Pressure in GPa
    T : float or array-like
        Temperature in °C

    Returns
    -------
    logfO2 : float or ndarray
    """
    P = np.asarray(P)
    T = np.asarray(T)

    T_kelvin = T + 273.15
    P_kbar = P * 10.0  # 1 GPa = 10 kbar

    numerator = (
        -587474.0
        + 1584.427 * T_kelvin
        - 203.3164 * T_kelvin * np.log(T_kelvin)
        + 0.09271 * T_kelvin**2
        + 1810.0 * P_kbar
    )

    denominator = 8.3144 * T_kelvin * np.log(10.0)

    return numerator / denominator


# =========================================================
# 对比函数
# 说明：
# - baseline 固定为 Frost (1991)
# - diff 使用“相对于 baseline 的差值”
#   即：model - baseline
# =========================================================
def compare_fmq(P=1.0, T_min=573, T_max=1200, n_points=300):
    """
    Compare three FMQ formulas using Frost (1991) as baseline.

    Parameters
    ----------
    P : float
        Pressure in GPa
    T_min : float
        Minimum temperature in °C
    T_max : float
        Maximum temperature in °C
    n_points : int
        Number of temperature points
    """
    T = np.linspace(T_min, T_max, n_points)

    fmq_frost = fmq_frost1991(P, T)
    fmq_oneill = fmq_oneill1987(P, T)
    fmq_hp = fmq_hp2011(P, T)

    # 以 Frost (1991) 为基线
    diff_oneill = fmq_oneill - fmq_frost
    diff_hp = fmq_hp - fmq_frost

    # 打印统计
    print(f"Pressure = {P:.3f} GPa")
    print(f"Temperature range = {T_min:.1f}–{T_max:.1f} °C")
    print("\nBaseline: Frost (1991)\n")

    print("O'Neill (1987a) - Frost (1991):")
    print(f"  Mean: {np.mean(diff_oneill):.4f}")
    print(f"  Std : {np.std(diff_oneill):.4f}")
    print(f"  Min : {np.min(diff_oneill):.4f}")
    print(f"  Max : {np.max(diff_oneill):.4f}")

    print("\nHolland & Powell - Frost (1991):")
    print(f"  Mean: {np.mean(diff_hp):.4f}")
    print(f"  Std : {np.std(diff_hp):.4f}")
    print(f"  Min : {np.min(diff_hp):.4f}")
    print(f"  Max : {np.max(diff_hp):.4f}")

    # -----------------------------
    # 图1：三个 FMQ 曲线
    # -----------------------------
    plt.figure(figsize=(8, 5))
    plt.plot(T, fmq_frost, label="FMQ (Frost 1991)", linewidth=2)
    plt.plot(T, fmq_oneill, label="FMQ (O'Neill 1987a)", linewidth=2)
    plt.plot(T, fmq_hp, label="FMQ (Holland & Powell)", linewidth=2)

    plt.xlabel("Temperature (°C)")
    plt.ylabel("log fO$_2$")
    plt.title(f"Comparison of FMQ formulas at {P} GPa")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # -----------------------------
    # 图2：相对 Frost (1991) 的差值
    # -----------------------------
    plt.figure(figsize=(8, 5))
    plt.plot(T, diff_oneill, label="O'Neill - Frost", linewidth=2)
    plt.plot(T, diff_hp, label="Holland & Powell - Frost", linewidth=2)

    plt.axhline(0, linestyle="--", linewidth=1, color="black")
    plt.xlabel("Temperature (°C)")
    plt.ylabel("Difference in log fO$_2$")
    plt.title(f"Difference relative to Frost (1991) at {P} GPa")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return {
        "T_C": T,
        "FMQ_Frost1991": fmq_frost,
        "FMQ_ONeill1987a": fmq_oneill,
        "FMQ_HollandPowell": fmq_hp,
        "diff_ONeill_minus_Frost": diff_oneill,
        "diff_HP_minus_Frost": diff_hp,
    }


# =========================================================
# 示例运行
# =========================================================
result = compare_fmq(P=2.0, T_min=573, T_max=1200, n_points=300)
```
